# ⚡ Smart Downloader
### Just run each cell top to bottom — it will ASK you for the URL


```python
# ═══════════════════════════════════
# CELL 1 — Install (Run ONCE only)
# ═══════════════════════════════════
import subprocess
print("Installing yt-dlp and ffmpeg via conda...")
result = subprocess.run(
    ["conda", "install", "-c", "conda-forge", "yt-dlp", "ffmpeg", "-y", "--no-update-deps"],
    capture_output=False
)
if result.returncode == 0:
    print("\n✅ Installed! Now run Cell 2.")
else:
    print("\n⚠️ Conda failed, trying pip...")
    subprocess.run(["pip", "install", "yt-dlp", "--break-system-packages"])
    print("✅ Done via pip. Now run Cell 2.")
```

```python
# ═══════════════════════════════════════════════════════════════
# CELL 2 — MAIN DOWNLOADER
# Just run this cell — it will ask you everything step by step
# ═══════════════════════════════════════════════════════════════

import os, yt_dlp
from datetime import datetime

# ─── STEP 1: Ask for URL ───
print("=" * 55)
print("  ⚡ SMART DOWNLOADER — Season / Movie / Playlist")
print("=" * 55)
print()
print("Supported sites: YouTube, Dailymotion, Archive.org,")
print("                 Vimeo, Bilibili, Tubi, and 1000+ more")
print()

while True:
    URL = input("📋 Paste your URL here and press Enter:\n> ").strip()
    if URL.startswith("http://") or URL.startswith("https://"):
        break
    print("❌ Invalid URL. Must start with http:// or https://  Try again.")

# ─── STEP 2: Ask for mode ───
print()
print("What do you want to download?")
print("  1 = Video / Full Season / Playlist  (saves as MP4)")
print("  2 = Audio only                      (saves as MP3)")
print()
while True:
    mode = input("Enter 1 or 2 [default=1]: ").strip() or "1"
    if mode in ("1", "2"):
        break
    print("❌ Enter 1 or 2")

# ─── STEP 3: Ask quality (video only) ───
quality = "3"
if mode == "1":
    print()
    print("Choose video quality:")
    print("  1 = 4K (2160p)       — very large files")
    print("  2 = 1440p            — large files")
    print("  3 = 1080p Full HD    — RECOMMENDED")
    print("  4 = 720p HD          — smaller files")
    print("  5 = 480p             — small files")
    print("  6 = 360p             — very small")
    print("  7 = Best available   — whatever is highest")
    print()
    while True:
        quality = input("Enter 1-7 [default=3 for 1080p]: ").strip() or "3"
        if quality in ("1","2","3","4","5","6","7"):
            break
        print("❌ Enter a number 1 to 7")

# ─── STEP 4: Ask folder name ───
print()
SHOW_NAME = input("📁 Folder name to save in (e.g. Breaking Bad S1)\n   Press Enter to use auto-name: ").strip()

# ─── Setup paths ───
base      = os.path.join(os.path.expanduser("~"), "downloads")
date_dir  = datetime.now().strftime("%Y-%m")
if SHOW_NAME:
    safe = "".join(c if c.isalnum() or c in " _-" else "_" for c in SHOW_NAME).strip()
else:
    safe = "misc"
folder  = os.path.join(base, date_dir, safe)
os.makedirs(folder, exist_ok=True)
archive = os.path.join(folder, "downloaded.txt")

print()
print("─" * 55)
print(f"🔗 URL     : {URL}")
print(f"📁 Folder  : {folder}")
print(f"📦 Archive : {archive}")
print(f"🎯 Quality : {quality if mode=='1' else 'Audio MP3'}")
print("─" * 55)
print()

# ─── Quality map ───
qmap = {
    "1": "bestvideo[height<=2160]+bestaudio/bestvideo+bestaudio/best",
    "2": "bestvideo[height<=1440]+bestaudio/bestvideo+bestaudio/best",
    "3": ("bestvideo[height<=1080]+bestaudio/"
          "bestvideo[height<=720]+bestaudio/"
          "bestvideo[height<=480]+bestaudio/best"),
    "4": "bestvideo[height<=720]+bestaudio/best",
    "5": "bestvideo[height<=480]+bestaudio/best",
    "6": "bestvideo[height<=360]+bestaudio/best",
    "7": "bestvideo+bestaudio/best",
}

# ─── Options ───
opts = {
    "outtmpl"            : os.path.join(folder, "%(playlist_index)s - %(title)s.%(ext)s"),
    "merge_output_format": "mp4",
    "ignoreerrors"       : True,       # skip broken videos, keep going
    "continuedl"         : True,       # resume if internet cuts
    "retries"            : 99,         # retry 99 times on error
    "fragment_retries"   : 99,
    "retry_sleep_functions": {"http": lambda n: min(5 * n, 30)},
    "download_archive"   : archive,    # never re-download finished files
    "nooverwrites"       : True,
    "extractor_args"     : {"youtube": {"player_client": ["android", "web"]}},
    "http_headers"       : {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/120.0.0.0 Safari/537.36"
        )
    },
}

if mode == "1":
    opts["format"] = qmap[quality]
else:
    opts["format"] = "bestaudio/best"
    opts["postprocessors"] = [{
        "key"             : "FFmpegExtractAudio",
        "preferredcodec"  : "mp3",
        "preferredquality": "192",
    }]

# ─── Download with cookie fallback ───
def run(use_cookies):
    o = opts.copy()
    if use_cookies:
        o["cookiesfrombrowser"] = ("chrome",)
    with yt_dlp.YoutubeDL(o) as ydl:
        ydl.download([URL])

print("⏬ Starting download...\n")
try:
    run(use_cookies=True)
except Exception as e:
    print(f"⚠️  Cookie mode failed → switching to public mode...\n   ({e})\n")
    try:
        run(use_cookies=False)
    except Exception as e2:
        print(f"❌ Download failed: {e2}")
        raise SystemExit

print()
print("✅ ALL DONE!")
print(f"📁 Saved in : {folder}")
print()
print("💡 If internet was cut — just run this cell AGAIN.")
print("   It will resume automatically and skip already-downloaded files.")
```

```python
# ═══════════════════════════════════════════════════════
# CELL 3 — OPTIONAL: See all available qualities & sizes
#           Run this before Cell 2 to pick best quality
# ═══════════════════════════════════════════════════════

import yt_dlp

print("🔍 Format Checker")
print("─" * 40)
URL = input("Paste URL to check formats:\n> ").strip()

if not URL.startswith("http"):
    print("❌ Invalid URL")
else:
    print(f"\n🔍 Loading...\n")
    with yt_dlp.YoutubeDL({"quiet": True, "skip_download": True}) as ydl:
        info = ydl.extract_info(URL, download=False)
        if info and "entries" in info:
            print(f"📋 Playlist: {info.get('title','')} — showing first video\n")
            info = next(iter(info["entries"]), None)
        if not info:
            print("❌ Could not get info.")
        else:
            print(f"🎬 {info.get('title','')}\n")
            print(f"{'ID':<12} {'EXT':<6} {'RESOLUTION':<12} {'FPS':<6} {'SIZE'}")
            print("─" * 55)
            seen = set()
            for f in info.get("formats", []):
                if f.get("vcodec","none") == "none": continue
                h   = f.get("height")
                res = f"{h}p" if h else "?"
                ext = f.get("ext","?")
                if (res, ext) in seen: continue
                seen.add((res, ext))
                fps  = f.get("fps") or "-"
                fs   = f.get("filesize") or f.get("filesize_approx")
                size = f"{fs/1024/1024:.1f} MB" if fs else "~unknown"
                fid  = f.get("format_id","?")
                print(f"{fid:<12} {ext:<6} {res:<12} {str(fps):<6} {size}")
            print("\n── Audio Only ──")
            print(f"{'ID':<12} {'EXT':<6} {'BITRATE':<12} {'SIZE'}")
            print("─" * 45)
            for f in info.get("formats", []):
                if f.get("vcodec","none") != "none": continue
                fid  = f.get("format_id","?")
                ext  = f.get("ext","?")
                abr  = f.get("abr") or "-"
                fs   = f.get("filesize") or f.get("filesize_approx")
                size = f"{fs/1024/1024:.1f} MB" if fs else "~unknown"
                print(f"{fid:<12} {ext:<6} {str(abr)+' kbps':<12} {size}")
            print("\n💡 Now run Cell 2 to download.")
```
