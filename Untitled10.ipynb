```python
#!/usr/bin/env python3
import os
import yt_dlp
from datetime import datetime

def is_valid_url(url: str) -> bool:
    return url.startswith("http")

def get_folder():
    path = f"downloads/{datetime.now().strftime('%Y-%m')}"
    os.makedirs(path, exist_ok=True)
    return path

def get_format(q="3"):
    return {
        "1": "bestvideo[height<=2160]+bestaudio/best",
        "2": "bestvideo[height<=1440]+bestaudio/best",
        "3": "bestvideo[height<=1080]+bestaudio/best",
        "4": "bestvideo[height<=720]+bestaudio/best",
        "5": "bestaudio/best"
    }.get(q, "bestvideo[height<=1080]+bestaudio/best")

def get_cookie_opts():
    print("""
--- Cookie Options (YouTube REQUIRES cookies now) ---
[1] Auto-extract from Chrome
[2] Auto-extract from Firefox
[3] Use a cookies.txt file  ← RECOMMENDED if 1/2 fail
[4] No cookies (will likely fail on YouTube)
""")
    choice = input("Cookie method: ").strip()

    if choice == "1":
        # Try common Chrome profile paths
        import platform
        system = platform.system()
        if system == "Windows":
            profiles = [
                None,  # default (let yt-dlp find it)
                os.path.expandvars(r"%LOCALAPPDATA%\Google\Chrome\User Data"),
                os.path.expandvars(r"%LOCALAPPDATA%\Google\Chrome\User Data\Default"),
            ]
        elif system == "Darwin":
            profiles = [
                None,
                os.path.expanduser("~/Library/Application Support/Google/Chrome"),
            ]
        else:  # Linux
            profiles = [
                None,
                os.path.expanduser("~/.config/google-chrome"),
                os.path.expanduser("~/.config/chromium"),
            ]

        for profile in profiles:
            if profile is None:
                return {"cookiesfrombrowser": ("chrome",)}
            elif os.path.exists(profile):
                print(f"   Using Chrome profile: {profile}")
                return {"cookiesfrombrowser": ("chrome", profile)}

        print("⚠️  Chrome profile not found. Try option 3 (cookies.txt) instead.")
        return get_cookie_opts()

    elif choice == "2":
        import platform
        system = platform.system()
        if system == "Windows":
            profiles = [
                None,
                os.path.expandvars(r"%APPDATA%\Mozilla\Firefox\Profiles"),
            ]
        else:
            profiles = [None]

        for profile in profiles:
            if profile is None:
                return {"cookiesfrombrowser": ("firefox",)}
            elif os.path.exists(profile):
                return {"cookiesfrombrowser": ("firefox", profile)}

        return {"cookiesfrombrowser": ("firefox",)}

    elif choice == "3":
        path = input("Path to cookies.txt file: ").strip().strip('"')
        if os.path.exists(path):
            print(f"✅ Using cookies file: {path}")
            return {"cookiefile": path}
        else:
            print(f"❌ File not found: {path}")
            print("   Make sure the path is correct. Try again.")
            return get_cookie_opts()

    else:
        print("⚠️  No cookies — YouTube will likely block this.")
        return {}

def download(url: str, mode="video", quality="3"):
    if not is_valid_url(url):
        print("❌ Invalid URL")
        return

    folder = get_folder()
    archive_file = "downloads/download_archive.txt"
    os.makedirs("downloads", exist_ok=True)

    ydl_opts = {
        "outtmpl": f"{folder}/%(playlist_title)s/%(playlist_index)s - %(title)s.%(ext)s",
        "noplaylist": False,
        "retries": 10,
        "fragment_retries": 10,
        "skip_unavailable_fragments": True,
        "quiet": False,
        "merge_output_format": "mp4",
        "download_archive": archive_file,  # ← skips already-downloaded videos
        "ignoreerrors": True,              # ← skip failed videos, don't stop entirely
        "sleep_interval": 2,               # ← small delay between videos (avoids rate limits)
        "max_sleep_interval": 5,
    }

    # Add cookies
    cookie_opts = get_cookie_opts()
    ydl_opts.update(cookie_opts)

    if mode == "audio":
        ydl_opts["format"] = "bestaudio/best"
        ydl_opts["postprocessors"] = [{
            "key": "FFmpegExtractAudio",
            "preferredcodec": "mp3",
            "preferredquality": "192",
        }]
    else:
        ydl_opts["format"] = get_format(quality)

    try:
        print(f"\n⏬ Starting download... (archive: {archive_file})")
        print("   Already-downloaded videos will be skipped automatically.\n")
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=True)

            if not info:
                print("❌ No data returned. Check your URL or cookies.")
                return

            entries = info.get("entries") or []
            # entries can contain None for skipped/failed items
            downloaded = [e for e in entries if e is not None]
            skipped = len(entries) - len(downloaded)

            print(f"\n✅ Done! {len(downloaded)} videos processed.")
            if skipped:
                print(f"   ({skipped} skipped — already downloaded or unavailable)")
            print(f"📁 Saved in: {folder}")
            print(f"📝 Archive: {archive_file}")

    except Exception as e:
        print(f"\n❌ Error: {str(e)}")
        print("\n💡 Tip: If you see 'Sign in to confirm you're not a bot',")
        print("   export cookies.txt from your browser and use option 3.")

def list_formats(url):
    cookie_opts = get_cookie_opts()
    opts = {"listformats": True}
    opts.update(cookie_opts)
    with yt_dlp.YoutubeDL(opts) as ydl:
        ydl.download([url])

def main():
    print("\n📝 Archive: downloads/download_archive.txt")
    print("   Re-run the same URL anytime — already-downloaded videos are skipped.\n")

    while True:
        print("""=========================
⚡ Downloader v9
=========================
[1] Video / Playlist Download
[2] Audio Download (MP3)
[3] List Formats
[0] Exit
""")
        c = input("Select: ").strip()
        if c == "0":
            break

        url = input("Paste URL: ").strip()

        if c == "1":
            q = input("Quality (1=4K / 2=1440 / 3=1080 / 4=720): ").strip()
            download(url, "video", q)
        elif c == "2":
            download(url, "audio")
        elif c == "3":
            list_formats(url)
        else:
            print("❌ Invalid choice")

if __name__ == "__main__":
    main()
```
