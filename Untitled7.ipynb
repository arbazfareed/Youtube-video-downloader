def _check_ffmpeg() -> bool:

    if _FFMPEG_OK is not None:

def _toggle_pause():
    global paused
    with pause_lock:
        paused = not paused
    state = "⏸  PAUSED   — press P to resume" if paused else "▶️  RESUMED  — downloading..."
    print(f"\n{state}\n")


def _keyboard_listener():
    """Daemon thread: press P to pause/resume, Q to quit."""
    try:
        if os.name == "nt":
            import msvcrt
            while True:
                if msvcrt.kbhit():
                    ch = msvcrt.getwch().lower()
                    if ch == "p":
                        _toggle_pause()
                    elif ch == "q":
                        print("\n👋  Quit requested.\n")
                        os._exit(0)
                time.sleep(0.1)
        else:
            import tty, termios, select
            fd  = sys.stdin.fileno()
            old = termios.tcgetattr(fd)
            try:
                tty.setraw(fd)
                while True:
                    r, _, _ = select.select([sys.stdin], [], [], 0.1)
                    if r:
                        ch = sys.stdin.read(1).lower()
                        if ch == "p":
                            _toggle_pause()
                        elif ch == "q":
                            print("\n👋  Quit requested.\n")
                            os._exit(0)
            finally:
                termios.tcsetattr(fd, termios.TCSADRAIN, old)
    except Exception:
        pass


def start_keyboard_listener():
    t = threading.Thread(target=_keyboard_listener, daemon=True)
    t.start()


def wait_if_paused():
    while True:
        with pause_lock:
            if not paused:
                return
        time.sleep(0.3)

def is_online(timeout=3) -> bool:
    """Try multiple hosts so one DNS failure doesn't lie to us."""
    for host, port in [("8.8.8.8", 53), ("1.1.1.1", 53), ("8.8.4.4", 53)]:
        try:
            s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            s.settimeout(timeout)
            s.connect((host, port))
            s.close()
            return True
        except OSError:
            continue
    return False


def wait_for_internet(retry_secs: int = 5) -> bool:
    """
    Block FOREVER until internet comes back.
    Returns True when reconnected.
    (We never give up — the user can Ctrl+C to abort.)
    """
    if is_online():
        return True
    print("\n🔌  Internet disconnected — waiting to reconnect (Ctrl+C to abort)...")
    elapsed = 0
    while True:
        for remaining in range(retry_secs, 0, -1):
            print(f"\r   Retrying in {remaining}s  (waited {elapsed}s total)   ",
                  end="", flush=True)
            time.sleep(1)
        elapsed += retry_secs
        if is_online():
            print("\n✅  Internet restored — resuming download!\n")
            return True

def load_history() -> list:
    if os.path.exists(HISTORY_FILE):
        try:
            with open(HISTORY_FILE, "r", encoding="utf-8") as f:
                return json.load(f)
        except Exception:
            pass
    return []


def save_history(entry: dict):
    history = load_history()
    history.append(entry)
    try:
        with open(HISTORY_FILE, "w", encoding="utf-8") as f:
            json.dump(history, f, indent=2, ensure_ascii=False)
    except Exception as e:
        print(f"⚠️  Could not save history: {e}")


def show_history():
    history = load_history()
    if not history:
        print("\n📭  No download history yet.")
        return
    print(f"\n📜  Download History  ({len(history)} total — showing last 30)")
    print("─" * 72)
    for i, e in enumerate(reversed(history[-30:]), 1):
        icon  = "✅" if e.get("status") == "ok" else "❌"
        ts    = e.get("time", "")[:16]
        title = (e.get("title") or e.get("url") or "?")[:52]
        print(f"  {i:>3}. {icon}  [{ts}]  {title}")
    print("─" * 72)


def already_downloaded(url: str) -> bool:
    for e in load_history():
        if e.get("url") == url and e.get("status") == "ok":
            return True
    return False

def make_folder(base: str, sub: str = None) -> str:
    now  = datetime.now()
    path = Path(base) / str(now.year) / f"{now.month:02d}"
    if sub:
        path /= sub
    path.mkdir(parents=True, exist_ok=True)
    return str(path)


def smart_folder(ctype: str) -> str:
    if ctype in ("video", "playlist", "audio"):
        sub = "videos" if ctype != "audio" else "audio"
        return make_folder("downloads", sub)
    return make_folder("downloads", "files")


def zip_folder(folder: str):
    zip_name = f"{folder.rstrip(os.sep)}.zip"
    with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zf:
        for root, _, files in os.walk(folder):
            for file in files:
                full = os.path.join(root, file)
                zf.write(full, os.path.relpath(full, folder))
    print(f"📦  Compressed → {zip_name}")


def detect_content_type(url: str) -> str:
    u = url.lower()
    if any(x in u for x in ["youtube", "youtu.be", "tiktok", "vimeo",
                              "facebook", "x.com", "twitter", "instagram", "reddit"]):
        return "video"
    if any(x in u for x in ["soundcloud", "bandcamp"]):
        return "audio"
    if u.endswith((".jpg", ".jpeg", ".png", ".gif", ".webp")):
        return "image"
    if u.endswith(".pdf"):
        return "pdf"
    if u.endswith((".zip", ".rar", ".7z")):
        return "archive"
    if u.endswith((".doc", ".docx", ".ppt", ".pptx", ".xls", ".xlsx")):
        return "document"
    # If URL contains common video platforms not caught above, still try yt-dlp
    return "video" if "://" in url else "unknown"

def build_format_string(resolution: str, audio_only: bool) -> str:
    """
    Build a robust format string that actually works for the chosen quality.

    KEY FIX: The old code only tried mp4+m4a combos which often 404.
    This version tries multiple codec combos and always has a fallback.
    """
    if audio_only:
        return "bestaudio[ext=m4a]/bestaudio[ext=opus]/bestaudio/best"

    if resolution == "best":
        if _check_ffmpeg():
            return "bestvideo+bestaudio/best"
        else:
            # No FFmpeg = can't merge, get best pre-merged stream
            return "best[ext=mp4]/best"

    h = resolution  # e.g. "1080"

    if _check_ffmpeg():
        # Try to get exactly the resolution requested, with multiple codec fallbacks
        return (
            f"bestvideo[height<={h}][ext=webm]+bestaudio[ext=webm]"
            f"/bestvideo[height<={h}][ext=mp4]+bestaudio[ext=m4a]"
            f"/bestvideo[height<={h}]+bestaudio"
            f"/best[height<={h}]"
            f"/bestvideo[height<=1080]+bestaudio"  # fallback to 1080 if exact not found
            f"/best"
        )
    else:
        # No FFmpeg — must use pre-merged streams only (no separate video+audio)
        print(f"  ℹ️  FFmpeg not found — downloading best pre-merged stream ≤{h}p")
        print(  "     Install FFmpeg for true quality selection.")
        return (
            f"best[height<={h}][ext=mp4]"
            f"/best[height<={h}]"
            f"/best[ext=mp4]"
            f"/best"
        )


def build_opts(folder: str,
               resolution:    str  = "1080",
               audio_only:    bool = False,
               audio_quality: str  = "192",
               proxy:         str  = None,
               limit_rate:    str  = None,
               subtitles:     bool = False,
               playlist:      bool = True) -> dict:

    fmt = build_format_string(resolution, audio_only)

    # Output template — safe for all platforms
    if audio_only:
        outtmpl = os.path.join(folder, "%(title)s.%(ext)s")
    else:
        outtmpl = os.path.join(folder, "%(title)s [%(height)sp].%(ext)s")

    opts = {
        # ── Format ─────────────────────────────────────────────────
        "format":              fmt,
        "merge_output_format": "mp4",
        "outtmpl":             outtmpl,

        # ── 403 / SABR / bot-detection fix ─────────────────────────
        "extractor_args": {
            "youtube": {
                "player_client": ["android", "web", "ios"],
            }
        },
        "http_headers": {
            "User-Agent": (
                "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                "AppleWebKit/537.36 (KHTML, like Gecko) "
                "Chrome/120.0.0.0 Safari/537.36"
            )
        },

        # ── Resilience — key settings ──────────────────────────────
        "retries":             999,       # ← INFINITE retries on error
        "fragment_retries":    999,       # ← INFINITE fragment retries
        "file_access_retries": 999,
        "continuedl":          True,      # NEVER restart from 0
        "keepvideo":           True,      # keep .part file on error
        "ignoreerrors":        False,     # we handle errors ourselves
        "noplaylist":          not playlist,
        "socket_timeout":      30,
        "retry_sleep_functions": {
            "http":     lambda n: min(4 * n, 30),
            "fragment": lambda n: min(2 * n, 10),
        },

        # ── Metadata ───────────────────────────────────────────────
        "writethumbnail":    True,
        "writeinfojson":     True,
        "writesubtitles":    subtitles,
        "writeautomaticsub": subtitles,
        "subtitleslangs":    ["en", "hi"] if subtitles else [],

        # ── Output ─────────────────────────────────────────────────
        "quiet":    False,
        "progress": True,
        "verbose":  False,

        # ── Post-processing ────────────────────────────────────────
        "postprocessors": [],
    }

    # Only add FFmpeg post-processors if FFmpeg is available
    if _check_ffmpeg():
        opts["postprocessors"] += [
            {"key": "FFmpegMetadata"},
        ]

    if proxy:
        opts["proxy"] = proxy
    if limit_rate:
        opts["ratelimit"] = limit_rate

    if audio_only:
        opts["postprocessors"].append({
            "key":              "FFmpegExtractAudio",
            "preferredcodec":   "mp3",
            "preferredquality": audio_quality,
        })

    return opts

# ════════════════════════════════════════════════════════════════
#  NETWORK ERROR DETECTION
# ════════════════════════════════════════════════════════════════
_NET_KEYWORDS = [
    "network is unreachable", "name or service not known",
    "failed to resolve", "errno -2", "errno 101", "errno 110",
    "timed out", "read timed out", "connection reset",
    "connection refused", "broken pipe", "remote end closed",
    "10054", "10060", "10061", "nodename nor servname",
    "temporary failure in name resolution", "no route to host",
    "network error", "ssl: eof", "ssl handshake",
    "handshake operation timed out",
]


def _is_network_error(text: str) -> bool:
    t = text.lower()
    return any(kw in t for kw in _NET_KEYWORDS)


# ════════════════════════════════════════════════════════════════
#  PROGRESS HOOK — pause-aware + stall detection
# ════════════════════════════════════════════════════════════════
def _make_progress_hook():
    last_bytes = [0]
    last_time  = [time.time()]
    stall_secs = 25   # seconds of 0 new bytes = treat as network drop

    def hook(d):
        wait_if_paused()

        if d["status"] == "downloading":
            downloaded = d.get("downloaded_bytes") or 0
            now        = time.time()

            if downloaded > last_bytes[0]:
                last_bytes[0] = downloaded
                last_time[0]  = now
            else:
                stalled = now - last_time[0]
                if stalled >= stall_secs and not is_online():
                    print(f"\n🔌  Download stalled {stalled:.0f}s — internet dropped")
                    wait_for_internet()
                    last_time[0] = time.time()

        elif d["status"] == "error":
            err = str(d.get("error", ""))
            if _is_network_error(err):
                print("\n🔌  Fragment download error — checking internet...")
                wait_for_internet()

    return hook


# ════════════════════════════════════════════════════════════════
#  CORE DOWNLOAD — bulletproof, infinite reconnect
# ════════════════════════════════════════════════════════════════
def _do_download(url: str, opts: dict) -> tuple:
    """
    Returns ("ok", title) on success, ("fail", error_message) on non-network failure.

    Never gives up on network errors — retries forever until internet returns.
    Uses continuedl=True so partial files are always resumed.
    """
    title = url

    # Inject our progress hook
    opts = dict(opts)  # don't mutate the caller's dict
    opts["progress_hooks"] = opts.get("progress_hooks", []) + [_make_progress_hook()]
    opts["continuedl"]     = True
    opts["retries"]        = 999
    opts["fragment_retries"] = 999

    attempt = 0
    while True:   # ← loop forever until success or fatal error
        wait_if_paused()

        if not is_online():
            print(f"\n🔌  No internet — waiting before attempt {attempt + 1}...")
            wait_for_internet()

        attempt += 1
        print(f"\n  ⏬  Download attempt #{attempt}  →  {url[:70]}\n")

        try:
            with yt_dlp.YoutubeDL(opts) as ydl:
                info = ydl.extract_info(url, download=True)
                if info:
                    title = info.get("title", url)
            print(f"\n✅  Download complete: {title[:70]}")
            return ("ok", title)

        except yt_dlp.utils.DownloadError as e:
            err = str(e)
            print(f"\n⚠️  yt-dlp error: {err[:120]}")

            if _is_network_error(err):
                print("🔌  Network error detected — waiting for internet...")
                wait_for_internet()
                print("    Resuming download from where it left off...\n")
                continue   # retry — continuedl will resume partial file

            # Non-network errors: try a few times then give up
            if attempt >= 3:
                print(f"\n❌  Non-network error after {attempt} attempts.")
                print("💡  Try:  pip install -U yt-dlp")
                print("💡  Or list formats first:  python downloader.py --formats <URL>")
                return ("fail", err)

            print(f"    Retrying ({attempt}/3)...")
            time.sleep(3)

        except KeyboardInterrupt:
            print("\n⛔  Cancelled by user.")
            return ("fail", "Cancelled by user")

        except Exception as e:
            err = str(e)
            print(f"\n⚠️  Unexpected error: {err[:120]}")
            if _is_network_error(err):
                wait_for_internet()
                continue
            if attempt >= 3:
                return ("fail", err)
            time.sleep(3)

#  SINGLE VIDEO / AUDIO DOWNLOAD
# ════════════════════════════════════════════════════════════════
def download_yt_content(url: str, content_type: str,
                        proxy: str = None, limit_rate: str = None):
    folder    = smart_folder(content_type)
    subtitles = ask_subtitles()

    if content_type == "audio":
        aq   = ask_audio_quality()
        opts = build_opts(folder, audio_only=True, audio_quality=aq,
                          proxy=proxy, limit_rate=limit_rate, subtitles=subtitles)
        res  = "audio"
        print(f"\n⏬  Downloading AUDIO  |  quality: {aq} kbps  |  folder: {folder}")
    else:
        if input("\n👁   Show available formats first? (y/n, default=n): ").strip().lower() == "y":
            list_formats(url)

        res  = ask_quality()
        aq   = ask_audio_quality()
        opts = build_opts(folder, resolution=res, audio_quality=aq,
                          proxy=proxy, limit_rate=limit_rate, subtitles=subtitles)

        if not _check_ffmpeg():
            print("\n  ⚠️  FFmpeg not found on your system.")
            print("     Without FFmpeg, video+audio cannot be merged.")
            print("     The script will download the best pre-merged stream instead.")
            print("     To get true quality selection, install FFmpeg:")
            print("       Windows: winget install Gyan.FFmpeg")
            print("       Linux:   sudo apt install ffmpeg")
            print("       macOS:   brew install ffmpeg\n")

        print(f"\n⏬  Downloading VIDEO  |  quality: {res}p  |  folder: {folder}")

    print("   Press  P  to pause/resume  |  Q  to quit")
    print("   Internet disconnects will be handled automatically\n")

    status, result = _do_download(url, opts)

    save_history({
        "time":       datetime.now().isoformat(timespec="seconds"),
        "url":        url,
        "title":      result if status == "ok" else url,
        "type":       content_type,
        "resolution": res,
        "status":     status,
        "folder":     folder,
    })

    if status == "ok":
        print(f"\n✅  Saved to: {folder}")
    else:
        print(f"\n❌  Failed: {result}")

#  GENERIC FILE DOWNLOAD  (PDF / image / document)
# ════════════════════════════════════════════════════════════════
def generic_file_download(url: str, max_retries: int = 9999):
    if not _HAS_REQUESTS:
        print("⚠️  requests library not installed.  Run:  pip install requests")
        return

    folder  = make_folder("downloads", "files")
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
    }
    print(f"\n📥  Downloading: {url}")

    attempt = 0
    while attempt < max_retries:
        wait_if_paused()
        if not is_online():
            wait_for_internet()

        attempt += 1
        try:
            r = requests.get(url, stream=True, timeout=30, headers=headers)
            if r.status_code != 200:
                print(f"❌  HTTP {r.status_code} — cannot download")
                return

            # Figure out filename
            ctype = r.headers.get("content-type", "")
            ext   = ctype.split("/")[-1].split(";")[0] if "/" in ctype else "bin"
            name  = url.split("/")[-1].split("?")[0]
            fname = safe_filename(name) if name else f"file_{int(time.time())}.{ext}"
            if "." not in fname:
                fname += f".{ext}"

            filepath = os.path.join(folder, fname)
            total    = int(r.headers.get("content-length", 0))
            done     = 0
            start    = time.time()

            with open(filepath, "ab") as f:
                for chunk in r.iter_content(65536):
                    wait_if_paused()
                    f.write(chunk)
                    done += len(chunk)
                    elapsed = max(time.time() - start, 0.001)
                    speed   = done / elapsed / 1024
                    if total:
                        pct = done / total * 100
                        print(f"\r  {pct:5.1f}%  {done//1024:,}KB/{total//1024:,}KB  {speed:.0f}KB/s  ",
                              end="", flush=True)
                    else:
                        print(f"\r  {done//1024:,}KB downloaded  {speed:.0f}KB/s  ",
                              end="", flush=True)

            print(f"\n✅  Saved → {filepath}")
            save_history({"time": datetime.now().isoformat(timespec="seconds"),
                          "url": url, "title": fname, "status": "ok", "folder": folder})
            return

        except requests.exceptions.ConnectionError:
            print(f"\n🔌  Connection lost (attempt {attempt}) — waiting...")
            wait_for_internet()
        except requests.exceptions.Timeout:
            print(f"\n⏱  Timeout (attempt {attempt}) — retrying...")
            time.sleep(3)
        except KeyboardInterrupt:
            print("\n⛔  Cancelled.")
            return
        except Exception as e:
            print(f"\n⚠️  Error: {e}")
            if attempt >= 5:
                print("❌  Giving up after 5 non-network errors.")
                return
            time.sleep(2)

    print("❌  Max retries reached.")


# ════════════════════════════════════════════════════════════════
#  BATCH URLS FILE PARSER
# ════════════════════════════════════════════════════════════════
def _parse_urls_file(file_path: str) -> list:
    """
    Supports:
      https://...             plain URL
      https://... # comment   URL with comment (ignored)
      https://... !!720       URL with quality override
      # comment line          skipped
    Returns list of (url, quality_or_None)
    """
    entries = []
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"):
                continue
            quality = None
            if " #" in line:
                line = line.split(" #")[0].strip()
            if " !!" in line:
                parts   = line.split(" !!", 1)
                line    = parts[0].strip()
                quality = parts[1].strip()
            if line:
                entries.append((line, quality))
    return entries


# ════════════════════════════════════════════════════════════════
#  BATCH MODE
# ════════════════════════════════════════════════════════════════
def _batch_worker(args):
    url, opts, ctype = args
    if ctype in ("video", "audio"):
        status, info = _do_download(url, opts)
        return (url, status, info)
    else:
        generic_file_download(url)
        return (url, "ok", url)


def batch_mode(file_path: str = "urls.txt"):
    if not os.path.exists(file_path):
        print(f"\n⚠️  File not found: {file_path}")
        print("  Create it with one URL per line.")
        print("  Examples:")
        print("    https://youtube.com/watch?v=abc")
        print("    https://youtube.com/watch?v=xyz  !!720   ← quality override")
        print("    # This is a comment — ignored")
        return

    entries = _parse_urls_file(file_path)
    if not entries:
        print("⚠️  No URLs found in the file.")
        return

    # Skip already-completed downloads
    fresh   = [(u, q) for u, q in entries if not already_downloaded(u)]
    skipped = len(entries) - len(fresh)
    if skipped:
        print(f"⏭   Skipping {skipped} already-downloaded URL(s).")
    if not fresh:
        print("✅  All URLs in the file have already been downloaded!")
        return

    print(f"\n📦  Batch mode — {len(fresh)} URL(s) to download  ({skipped} skipped)\n")

    default_res = ask_quality()
    aq          = ask_audio_quality()
    subtitles   = ask_subtitles()
    proxy       = ask_proxy()
    limit_rate  = ask_speed_limit()
    workers     = ask_workers()

    tasks = []
    for url, quality_override in fresh:
        ctype  = detect_content_type(url)
        folder = smart_folder(ctype)
        res    = quality_override or default_res
        opts   = build_opts(folder, resolution=res,
                            audio_only=(ctype == "audio"),
                            audio_quality=aq, proxy=proxy,
                            limit_rate=limit_rate, subtitles=subtitles)
        tasks.append((url, opts, ctype))

    print(f"\n⚡  Starting {workers} parallel worker(s)...")
    print("   Press  P  to pause/resume  |  Q  to quit\n")

    ok_count = fail_count = 0
    with ThreadPoolExecutor(max_workers=workers) as pool:
        futures = {pool.submit(_batch_worker, t): t[0] for t in tasks}
        for future in as_completed(futures):
            url = futures[future]
            try:
                _, status, info = future.result()
                if status == "ok":
                    ok_count += 1
                    print(f"  ✅  {str(info)[:65]}")
                else:
                    fail_count += 1
                    print(f"  ❌  {url[:50]}  →  {str(info)[:45]}")
                save_history({
                    "time":   datetime.now().isoformat(timespec="seconds"),
                    "url":    url,
                    "title":  str(info),
                    "status": status,
                })
            except Exception as e:
                fail_count += 1
                print(f"  ⚠️  {url[:50]}  →  {e}")

    print(f"\n═══ Batch complete ═══  ✅ {ok_count} succeeded   ❌ {fail_count} failed")
    if input("\n📦  Compress all downloads to ZIP? (y/n): ").strip().lower() == "y":
        zip_folder("downloads")


# ════════════════════════════════════════════════════════════════
#  CLIPBOARD MONITOR MODE
# ════════════════════════════════════════════════════════════════
VIDEO_DOMAINS = [
    "youtube.com", "youtu.be", "tiktok.com", "vimeo.com",
    "facebook.com", "instagram.com", "twitter.com", "x.com",
    "reddit.com", "soundcloud.com", "bandcamp.com",
    "dailymotion.com", "twitch.tv", "rumble.com",
]


def _get_clipboard() -> str:
    try:
        if os.name == "nt":
            r = subprocess.run(["powershell", "-command", "Get-Clipboard"],
                               capture_output=True, text=True, timeout=3)
            return r.stdout.strip()
        else:
            for cmd in [["xclip", "-selection", "clipboard", "-o"],
                        ["xsel", "--clipboard", "--output"],
                        ["wl-paste", "--no-newline"],
                        ["pbpaste"]]:
                try:
                    r = subprocess.run(cmd, capture_output=True, text=True, timeout=3)
                    if r.returncode == 0:
                        return r.stdout.strip()
                except FileNotFoundError:
                    continue
    except Exception:
        pass
    return ""


def _looks_like_media_url(text: str) -> bool:
    if not text.startswith(("http://", "https://")):
        return False
    return any(d in text.lower() for d in VIDEO_DOMAINS)


def clipboard_monitor_mode():
    print("\n📋  CLIPBOARD MONITOR — running")
    print("    Copy any video/audio URL and it will be auto-detected.")
    print("    Press  Ctrl+C  to stop.\n")
    seen      = set()
    last_text = _get_clipboard()
    try:
        while True:
            time.sleep(1)
            text = _get_clipboard()
            if text and text != last_text and text not in seen:
                last_text = text
                if _looks_like_media_url(text):
                    seen.add(text)
                    print(f"\n🔗  Detected: {text[:80]}")
                    ctype = detect_content_type(text)
                    print(f"🧠  Type: {ctype.upper()}")
                    confirm = input("⬇️   Download? (Enter=yes / n=skip): ").strip().lower()
                    if confirm != "n":
                        download_yt_content(text, ctype)
                    print("\n📋  Watching clipboard...")
    except KeyboardInterrupt:
        print("\n⏹  Clipboard monitor stopped.")


# ════════════════════════════════════════════════════════════════
#  WATCH FOLDER MODE
# ════════════════════════════════════════════════════════════════
def watch_folder_mode():
    queue_dir = Path(WATCH_FOLDER)
    done_dir  = queue_dir / "done"
    queue_dir.mkdir(exist_ok=True)
    done_dir.mkdir(exist_ok=True)

    print(f"\n📂  WATCH FOLDER MODE — monitoring: {queue_dir.resolve()}")
    print("    Drop a .txt file with one URL per line to auto-download.")
    print("    Processed files move to  watch_queue/done/")
    print("    Press  Ctrl+C  to stop.\n")

    processed = set()
    try:
        while True:
            time.sleep(2)
            for txt_file in queue_dir.glob("*.txt"):
                if txt_file in processed:
                    continue
                processed.add(txt_file)
                print(f"\n📄  New file: {txt_file.name}")
                try:
                    urls = [l.strip() for l in txt_file.read_text(encoding="utf-8").splitlines()
                            if l.strip() and not l.startswith("#")]
                except Exception as e:
                    print(f"⚠️  Cannot read: {e}")
                    continue

                if not urls:
                    print("   (empty — skipping)")
                else:
                    print(f"   {len(urls)} URL(s) found — starting...\n")
                    res = ask_quality()
                    aq  = ask_audio_quality()
                    for url in urls:
                        ctype  = detect_content_type(url)
                        folder = smart_folder(ctype)
                        opts   = build_opts(folder, resolution=res,
                                            audio_only=(ctype == "audio"),
                                            audio_quality=aq)
                        status, result = _do_download(url, opts)
                        icon = "✅" if status == "ok" else "❌"
                        print(f"  {icon}  {str(result)[:65]}")
                        save_history({"time": datetime.now().isoformat(timespec="seconds"),
                                      "url": url, "title": str(result), "status": status})

                txt_file.rename(done_dir / txt_file.name)
                print(f"\n📁  Moved to done/")
                print("📂  Watching...")
    except KeyboardInterrupt:
        print("\n⏹  Watch folder stopped.")

def _is_jupyter() -> bool:
    try:
        shell = get_ipython().__class__.__name__  # type: ignore
        return shell in ("ZMQInteractiveShell", "TerminalInteractiveShell")
    except NameError:
        return False


def main():
    start_keyboard_listener()

    if _is_jupyter():
        print("📓  Jupyter detected — opening interactive menu.\n")
        interactive_menu()
        return

    args = sys.argv[1:]
    if not args:
        interactive_menu()
        return

    cmd = args[0]

    if cmd == "--batch":
        batch_mode(args[1] if len(args) > 1 else "urls.txt")
    elif cmd == "--clipboard":
        clipboard_monitor_mode()
    elif cmd == "--watch":
        watch_folder_mode()
    elif cmd == "--history":
        show_history()
    elif cmd == "--formats":
        url = args[1] if len(args) > 1 else input("🔗  URL: ").strip()
        list_formats(url)
        elif cmd == "--help":8
        print(__doc__)
    elif cmd.startswith("http"):
        url   = cmd
        ctype = detect_content_type(url)
        print(f"🧠  Detected type: {ctype.upper()}")
        if ctype in ("video", "audio", "unknown"):
            download_yt_content(url, ctype if ctype != "unknown" else "video")
        else:
            generic_file_download(url)
    else:
        print(f"⚠️  Unknown argument '{cmd}' — opening menu instead.")
        interactive_menu()


if __name__ == "__main__":
    main()
