````python
# ▶ CELL 2 — Imports & Setup (run once per session)
import os
import yt_dlp
from datetime import datetime
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

print("✅ All imports loaded successfully.")
````

````python
# ▶ CELL 3 — Core Functions (run once)

def is_valid_url(url):
    return url.strip().startswith("http://") or url.strip().startswith("https://")

def get_folder(custom_name=None):
    base = os.path.join(os.path.expanduser("~"), "downloads")
    date_folder = datetime.now().strftime("%Y-%m")
    if custom_name and custom_name.strip():
        safe = "".join(c if c.isalnum() or c in " _-" else "_" for c in custom_name).strip()
        path = os.path.join(base, date_folder, safe)
    else:
        path = os.path.join(base, date_folder)
    os.makedirs(path, exist_ok=True)
    return path

def smart_format(q):
    quality_map = {
        "1": "bestvideo[height<=2160]+bestaudio/bestvideo+bestaudio/best",
        "2": "bestvideo[height<=1440]+bestaudio/bestvideo+bestaudio/best",
        "3": ("bestvideo[height<=1080]+bestaudio/"
              "bestvideo[height<=720]+bestaudio/"
              "bestvideo[height<=480]+bestaudio/best"),
        "4": "bestvideo[height<=720]+bestaudio/best",
        "5": "bestvideo[height<=480]+bestaudio/best",
        "6": "bestvideo[height<=360]+bestaudio/best",
        "7": "bestvideo+bestaudio/best",
    }
    return quality_map.get(q, q)

print("✅ Core functions ready.")
````

````python
# ▶ CELL 4 — 🔍 List ALL Available Formats & File Sizes
# Change this URL to your video/playlist link
URL = "https://www.youtube.com/watch?v=EXAMPLE"   # ← paste your URL here

# ──────────────────────────────────────────────
if not is_valid_url(URL):
    print("❌ Invalid URL — paste a real link above")
else:
    print(f"🔍 Fetching formats for:\n   {URL}\n")
    ydl_opts = {"quiet": True, "no_warnings": True, "skip_download": True}

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(URL, download=False)

        if "entries" in info:
            print(f"📋 Playlist: {info.get('title','Unknown')} — showing formats for first video\n")
            info = next(iter(info["entries"]), None)

        if not info:
            print("❌ Could not extract video info.")
        else:
            print(f"🎬 {info.get('title','Unknown')}\n")
            print(f"{'ID':<12} {'EXT':<6} {'RESOLUTION':<12} {'FPS':<6} {'SIZE':<12} NOTE")
            print("─" * 68)

            seen = set()
            for f in info.get("formats", []):
                if f.get("vcodec","none") == "none":
                    continue
                height   = f.get("height")
                res      = f"{height}p" if height else "?"
                ext      = f.get("ext","?")
                if (res, ext) in seen:
                    continue
                seen.add((res, ext))
                fps      = f.get("fps") or "-"
                note     = f.get("format_note","")
                fmt_id   = f.get("format_id","?")
                fs       = f.get("filesize") or f.get("filesize_approx")
                size_mb  = f"{fs/(1024*1024):.1f} MB" if fs else "~unknown"
                print(f"{fmt_id:<12} {ext:<6} {res:<12} {str(fps):<6} {size_mb:<12} {note}")

            print("\n── Audio Only ──")
            print(f"{'ID':<12} {'EXT':<6} {'BITRATE':<12} {'SIZE'}")
            print("─" * 45)
            for f in info.get("formats", []):
                if f.get("vcodec","none") != "none":
                    continue
                fmt_id  = f.get("format_id","?")
                ext     = f.get("ext","?")
                abr     = f.get("abr") or "-"
                fs      = f.get("filesize") or f.get("filesize_approx")
                size_mb = f"{fs/(1024*1024):.1f} MB" if fs else "~unknown"
                print(f"{fmt_id:<12} {ext:<6} {str(abr)+' kbps':<12} {size_mb}")

            print("\n💡 Copy a format ID above and paste into CELL 6 if you want exact quality.")
````

````python
# ▶ CELL 5 — ⬇️ Download VIDEO / Full Season / Playlist
# ════════════════════════════════════════════════
URL        = "https://www.youtube.com/playlist?list=XXXX"  # ← your URL
SHOW_NAME  = "My Show Season 1"   # ← folder name (leave "" to auto-name)
QUALITY    = "3"                  # ← 1=4K 2=1440p 3=1080p 4=720p 5=480p 6=360p 7=best
#              OR paste a format ID from CELL 4 e.g. "137+140"
# ════════════════════════════════════════════════

if not is_valid_url(URL):
    print("❌ Invalid URL")
else:
    folder       = get_folder(SHOW_NAME)
    archive_file = os.path.join(folder, "downloaded.txt")

    print(f"📁 Save folder : {folder}")
    print(f"📦 Archive file: {archive_file}")
    print(f"🎯 Quality     : {QUALITY}\n")

    out_tmpl = os.path.join(folder,
        "%(playlist_title)s",
        "%(playlist_index)s - %(title)s.%(ext)s")

    base_opts = {
        "outtmpl"            : out_tmpl,
        "ignoreerrors"       : True,
        "continuedl"         : True,
        "retries"            : 20,
        "fragment_retries"   : 20,
        "nooverwrites"       : True,
        "download_archive"   : archive_file,
        "merge_output_format": "mp4",
        "format"             : smart_format(QUALITY),
        "extractor_args"     : {"youtube": {"player_client": ["android","web"]}},
        "http_headers"       : {
            "User-Agent": ("Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                           "AppleWebKit/537.36 (KHTML, like Gecko) "
                           "Chrome/120.0.0.0 Safari/537.36")
        },
    }

    def run(use_cookies):
        opts = base_opts.copy()
        if use_cookies:
            opts["cookiesfrombrowser"] = ("chrome",)
        with yt_dlp.YoutubeDL(opts) as ydl:
            ydl.download([URL])

    try:
        print("⏬ Trying with Chrome cookies...\n")
        run(True)
    except Exception as e:
        print(f"⚠️  Cookie mode failed: {e}\n⏬ Retrying without cookies...")
        try:
            run(False)
        except Exception as e2:
            print(f"❌ Failed: {e2}")

    print(f"\n✅ DONE!  Files saved in: {folder}")
````

````python
# ▶ CELL 6 — 🎵 Download AUDIO (MP3)
# ════════════════════════════════════════════════
URL       = "https://www.youtube.com/watch?v=EXAMPLE"  # ← your URL
SHOW_NAME = "My Playlist"   # ← folder name
# ════════════════════════════════════════════════

if not is_valid_url(URL):
    print("❌ Invalid URL")
else:
    folder       = get_folder(SHOW_NAME)
    archive_file = os.path.join(folder, "downloaded.txt")

    print(f"📁 Saving MP3s to: {folder}\n")

    opts = {
        "outtmpl"         : os.path.join(folder, "%(title)s.%(ext)s"),
        "format"          : "bestaudio/best",
        "ignoreerrors"    : True,
        "continuedl"      : True,
        "retries"         : 20,
        "nooverwrites"    : True,
        "download_archive": archive_file,
        "postprocessors"  : [{
            "key"             : "FFmpegExtractAudio",
            "preferredcodec"  : "mp3",
            "preferredquality": "192",
        }],
    }

    with yt_dlp.YoutubeDL(opts) as ydl:
        ydl.download([URL])

    print(f"\n✅ Audio saved in: {folder}")
````

## 📖 How to Find Season / Show Links

| Source | Best For | Example URL Pattern |
|--------|----------|---------------------|
| **YouTube** | Official uploads, anime, classic TV | `youtube.com/playlist?list=PLxxxxx` |
| **Dailymotion** | TV episodes, international shows | `dailymotion.com/playlist/xxxxx` |
| **Archive.org** | Classic/public domain shows | `archive.org/details/show-name` |
| **Bilibili** | Anime, Asian dramas | `bilibili.com/video/BVxxxxx` |
| **Vimeo** | Indie films, documentaries | `vimeo.com/channels/name` |
| **Crunchyroll** | Anime (free episodes) | `crunchyroll.com/series/xxx` |

### Steps to get a playlist URL:
1. Open the show/season page in your browser  
2. Click on the **Playlist** tab (on YouTube)  
3. Copy the full URL from the address bar  
4. Paste it into **CELL 5** above  

### 💡 Recommended Workflow:
```
CELL 4 → Paste URL → See all sizes/formats
CELL 5 → Paste URL → Pick quality → Download full season
```
