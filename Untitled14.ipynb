```python
#!/usr/bin/env python3
import os
import yt_dlp
from datetime import datetime

# -----------------------------
# VALIDATION
# -----------------------------

def is_valid_url(url: str) -> bool:
    return url.startswith("http")

def is_playlist(url: str) -> bool:
    return "playlist?list=" in url or "&list=" in url

# -----------------------------
# FOLDER SYSTEM
# -----------------------------

def get_folder():
    path = f"downloads/{datetime.now().strftime('%Y-%m')}"
    os.makedirs(path, exist_ok=True)
    return path

# -----------------------------
# SMART QUALITY FALLBACK
# -----------------------------

def smart_format(q="3"):
    """
    Priority fallback:
    1080 -> 720 -> 480 -> best available
    """
    if q == "1":  # 4K
        return "bestvideo[height<=2160]+bestaudio/bestvideo+bestaudio/best"
    if q == "2":  # 1440p
        return "bestvideo[height<=1440]+bestaudio/bestvideo+bestaudio/best"
    if q == "3":  # 1080p (SMART fallback)
        return (
            "bestvideo[height<=1080]+bestaudio/"
            "bestvideo[height<=720]+bestaudio/"
            "bestvideo[height<=480]+bestaudio/"
            "best"
        )
    if q == "4":  # 720p
        return "bestvideo[height<=720]+bestaudio/best"
    return "best"

# -----------------------------
# MAIN DOWNLOAD ENGINE
# -----------------------------

def download(url: str, mode="video", quality="3"):
    if not is_valid_url(url):
        print("❌ Invalid URL")
        return

    folder = get_folder()
    archive_file = os.path.join(folder, "downloaded.txt")

    base_opts = {
        "outtmpl": f"{folder}/%(playlist_title)s/%(playlist_index)s - %(title)s.%(ext)s",

        # 🔥 IMPORTANT FIXES
        "ignoreerrors": True,          # skip broken videos
        "continuedl": True,           # resume broken downloads
        "retries": 20,
        "fragment_retries": 20,
        "nooverwrites": True,
        "download_archive": archive_file,  # skip already downloaded

        "merge_output_format": "mp4",

        # 🔥 IMPROVED EXTRACTION (fix bot issues)
        "extractor_args": {
            "youtube": {
                "player_client": ["android", "web"]
            }
        }
    }

    # -----------------------------
    # COOKIE HANDLING (Chrome + fallback)
    # -----------------------------

    def run(use_cookies=True):
        ydl_opts = base_opts.copy()

        if use_cookies:
            ydl_opts["cookiesfrombrowser"] = ("chrome",)

        if mode == "audio":
            ydl_opts["format"] = "bestaudio/best"
            ydl_opts["postprocessors"] = [{
                "key": "FFmpegExtractAudio",
                "preferredcodec": "mp3",
                "preferredquality": "192",
            }]
        else:
            ydl_opts["format"] = smart_format(quality)

        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            return ydl.extract_info(url, download=True)

    # -----------------------------
    # EXECUTION FLOW
    # -----------------------------

    try:
        print("\n⏬ Starting SMART download (Chrome cookies)...\n")
        run(use_cookies=True)

    except Exception as e:
        print("\n⚠ Chrome cookies failed → fallback mode...\n")
        print("Reason:", e)

        try:
            print("\n⏬ Retrying WITHOUT cookies (public mode)...\n")
            run(use_cookies=False)

        except Exception as e2:
            print("\n❌ COMPLETE FAILURE:", e2)
            return

    print("\n✅ DONE (auto-resume + smart fallback enabled)")
    print("📁 Saved in:", folder)

# -----------------------------
# FORMAT CHECK (optional)
# -----------------------------

def list_formats(url):
    with yt_dlp.YoutubeDL({"listformats": True}) as ydl:
        ydl.download([url])

# -----------------------------
# MENU
# -----------------------------

def main():
    while True:
        print("""
=========================
⚡ SMART DOWNLOADER v8
=========================
[1] Video / Playlist Download
[2] Audio Download (MP3)
[3] List Formats
[0] Exit
""")

        choice = input("Select: ").strip()

        if choice == "0":
            break

        url = input("Paste URL: ").strip()

        if choice == "1":
            print("""
Quality:
1 = 4K
2 = 1440p
3 = 1080p (SMART fallback)
4 = 720p
""")
            q = input("Choose quality: ").strip()
            download(url, "video", q)

        elif choice == "2":
            download(url, "audio")

        elif choice == "3":
            list_formats(url)

        else:
            print("❌ Invalid choice")

if __name__ == "__main__":
    main()
```
