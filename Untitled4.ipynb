```python
import yt_dlp
import requests
import os
import sys
import time
from pathlib import Path
from datetime import datetime
import zipfile

# ───────────────────────────────
#  Utility Functions
# ───────────────────────────────
def safe_filename(name):
    for char in '\\/*?:"<>|':
        name = name.replace(char, "_")
    return name.strip()


def make_folder(base, sub=None):
    now = datetime.now()
    path = Path(base) / str(now.year) / f"{now.month:02d}"
    if sub:
        path /= sub
    path.mkdir(parents=True, exist_ok=True)
    return str(path)


def zip_folder(folder):
    zip_name = f"{folder.rstrip(os.sep)}.zip"
    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(folder):
            for file in files:
                full = os.path.join(root, file)
                zipf.write(full, os.path.relpath(full, folder))
    print(f"📦 Downloads compressed → {zip_name}")


def detect_content_type(url):
    url_l = url.lower()
    if any(x in url_l for x in ["youtube", "youtu.be", "tiktok", "vimeo", "facebook", "x.com", "twitter", "instagram", "reddit"]):
        return "video"
    if any(x in url_l for x in ["soundcloud", "bandcamp", "spotify"]):
        return "audio"
    if url_l.endswith((".jpg", ".jpeg", ".png", ".gif", ".webp")):
        return "image"
    if url_l.endswith(".pdf"):
        return "pdf"
    if url_l.endswith((".zip", ".rar", ".7z")):
        return "archive"
    if url_l.endswith((".doc", ".docx", ".ppt", ".pptx", ".xls", ".xlsx")):
        return "document"
    return "unknown"


def smart_download_folder(content_type):
    if content_type in ["video", "playlist"]:
        return make_folder("downloads", "videos")
    if content_type == "audio":
        return make_folder("downloads", "audio")
    return make_folder("downloads", "files")


# ───────────────────────────────
#  Quality Selection Menu
# ───────────────────────────────
QUALITY_MAP = {
    "1": ("2160", "4K Ultra HD (2160p)"),
    "2": ("1440", "2K / QHD (1440p)"),
    "3": ("1080", "Full HD (1080p)  ← Recommended"),
    "4": ("720",  "HD Ready (720p)"),
    "5": ("480",  "SD (480p)"),
    "6": ("360",  "Low (360p)"),
    "7": ("best", "Best available (auto)"),
}

def ask_quality():
    print("\n🎬 Select video quality:")
    print("─────────────────────────────")
    for key, (_, label) in QUALITY_MAP.items():
        print(f"  [{key}] {label}")
    print("─────────────────────────────")
    choice = input("Enter number (default = 3 for 1080p): ").strip() or "3"
    res, label = QUALITY_MAP.get(choice, ("1080", "Full HD (1080p)"))
    print(f"✅ Selected: {label}")
    return res


def ask_audio_format():
    print("\n🎵 Select audio quality (for audio-only or merged):")
    print("  [1] 320 kbps (Best)")
    print("  [2] 192 kbps  ← Recommended")
    print("  [3] 128 kbps")
    print("  [4] 64 kbps (Small size)")
    choice = input("Enter number (default = 2): ").strip() or "2"
    quality_map = {"1": "320", "2": "192", "3": "128", "4": "64"}
    return quality_map.get(choice, "192")


# ───────────────────────────────
#  List Available Formats
# ───────────────────────────────
def list_formats(url):
    """Show all available formats for the URL."""
    print(f"\n📋 Fetching available formats for: {url}\n")
    opts = {
        "listformats": True,
        "quiet": False,
        # FIX: Use android client to bypass 403 issues
        "extractor_args": {"youtube": {"player_client": ["android", "web"]}},
    }
    try:
        with yt_dlp.YoutubeDL(opts) as ydl:
            ydl.download([url])
    except Exception as e:
        print(f"⚠️ Could not list formats: {e}")


# ───────────────────────────────
#  Build yt-dlp Options  (FIXED)
# ───────────────────────────────
def build_opts(folder, resolution="1080", audio_only=False, audio_quality="192", proxy=None, limit_rate=None):
    """
    KEY FIXES vs old code:
    1. extractor_args  → forces android+web clients, avoids 403 / SABR blocking
    2. format string   → picks the requested resolution with audio fallback
    3. Removed FFmpegAudioVolumeNormalize (not always available, caused errors)
    4. cookiesfrombrowser → optional, helps if still blocked
    """

    # ── Format string ──────────────────────────────────────────────
    if audio_only:
        fmt = "bestaudio/best"
    elif resolution == "best":
        fmt = "bestvideo+bestaudio/best"
    else:
        # Try exact resolution, then fall back to next best
        fmt = (
            f"bestvideo[height<={resolution}][ext=mp4]+bestaudio[ext=m4a]"
            f"/bestvideo[height<={resolution}]+bestaudio"
            f"/best[height<={resolution}]"
            f"/bestvideo+bestaudio/best"
        )

    opts = {
        # ── Output ────────────────────────────────────────────────
        "outtmpl": os.path.join(folder, "%(title)s [%(height)sp].%(ext)s"),
        "merge_output_format": "mp4",
        "format": fmt,

        # ── FIX 1: Force android client → avoids 403 / SABR error ─
        "extractor_args": {
            "youtube": {
                "player_client": ["android", "web"],
            }
        },

        # ── Resilience ────────────────────────────────────────────
        "retries": 10,
        "fragment_retries": 10,
        "continuedl": True,
        "ignoreerrors": True,
        "noplaylist": False,

        # ── Metadata ──────────────────────────────────────────────
        "writethumbnail": True,
        "writeinfojson": True,
        "writesubtitles": False,        # disabled by default (live_chat was huge)
        "write_description": True,

        # ── Progress ──────────────────────────────────────────────
        "quiet": False,
        "progress": True,

        # ── Post-processing ───────────────────────────────────────
        "postprocessors": [
            {"key": "FFmpegMetadata"},
            {"key": "EmbedThumbnail"},   # embed thumbnail into file
        ],
    }

    if proxy:
        opts["proxy"] = proxy
    if limit_rate:
        opts["ratelimit"] = limit_rate

    if audio_only:
        opts["postprocessors"].append({
            "key": "FFmpegExtractAudio",
            "preferredcodec": "mp3",
            "preferredquality": audio_quality,
        })

    return opts


# ───────────────────────────────
#  yt-dlp Download
# ───────────────────────────────
def download_yt_content(url, content_type):
    folder = smart_download_folder(content_type)

    if content_type == "audio":
        resolution = None
        audio_quality = ask_audio_format()
        opts = build_opts(folder, audio_only=True, audio_quality=audio_quality)
    else:
        # Ask quality
        show_formats = input("\n👀 Show all available formats first? (y/n, default=n): ").strip().lower()
        if show_formats == "y":
            list_formats(url)

        resolution = ask_quality()
        audio_quality = "192"
        opts = build_opts(folder, resolution=resolution, audio_only=False, audio_quality=audio_quality)

    try:
        print(f"\n⏬  Downloading {content_type.upper()} → {resolution or 'audio'}  |  folder: {folder}\n")
        with yt_dlp.YoutubeDL(opts) as ydl:
            ydl.download([url])
        print(f"\n✅  Done! Saved in: {folder}")
    except yt_dlp.utils.DownloadError as e:
        print(f"\n❌  Download failed: {e}")
        print("💡  Tips:")
        print("    • Update yt-dlp:  pip install -U yt-dlp")
        print("    • Try with cookies: add --cookies-from-browser chrome in CLI")
        print("    • Try a different quality option")
    except Exception as e:
        print(f"\n⚠️  Unexpected error: {e}")


# ───────────────────────────────
#  Generic File Downloader
# ───────────────────────────────
def generic_file_download(url):
    folder = smart_download_folder("files")
    print(f"📥 Downloading file from: {url}")
    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        r = requests.get(url, stream=True, timeout=20, headers=headers)
        if r.status_code != 200:
            print(f"❌ Server returned HTTP {r.status_code}")
            return
        content_type = r.headers.get("content-type", "")
        ext = content_type.split('/')[-1].split(';')[0] if '/' in content_type else 'bin'
        filename = safe_filename(url.split('/')[-1].split('?')[0]) or f"file_{int(time.time())}.{ext}"
        filepath = os.path.join(folder, filename)
        total = int(r.headers.get("content-length", 0))
        downloaded = 0
        with open(filepath, "wb") as f:
            for chunk in r.iter_content(8192):
                f.write(chunk)
                downloaded += len(chunk)
                if total:
                    pct = downloaded / total * 100
                    print(f"\r  Progress: {pct:.1f}%", end="", flush=True)
        print(f"\n✅ Saved → {filepath}")
    except Exception as e:
        print(f"⚠️ Error: {e}")


# ───────────────────────────────
#  Batch Mode
# ───────────────────────────────
def batch_mode(file_path="urls.txt"):
    if not os.path.exists(file_path):
        print("⚠️  Create a urls.txt file with one URL per line.")
        return
    with open(file_path, "r") as f:
        urls = [line.strip() for line in f if line.strip() and not line.startswith("#")]

    print(f"📦 Batch mode — {len(urls)} URLs found\n")

    # Ask quality once for all
    resolution = ask_quality()
    audio_quality = ask_audio_format()

    for i, url in enumerate(urls, 1):
        print(f"\n[{i}/{len(urls)}] {url}")
        ctype = detect_content_type(url)
        if ctype in ["video", "audio"]:
            folder = smart_download_folder(ctype)
            opts = build_opts(
                folder,
                resolution=resolution,
                audio_only=(ctype == "audio"),
                audio_quality=audio_quality,
            )
            try:
                with yt_dlp.YoutubeDL(opts) as ydl:
                    ydl.download([url])
            except Exception as e:
                print(f"  ⚠️ Failed: {e}")
        else:
            generic_file_download(url)

    compress = input("\n📦 Compress all downloads into ZIP? (y/n): ").lower()
    if compress == "y":
        zip_folder("downloads")
    print("\n✅ Batch complete.")


# ───────────────────────────────
#  Main Entry
# ───────────────────────────────
def main():
    print("⚡  Universal Downloader X — Ultimate Edition  v2.0")
    print("════════════════════════════════════════════════════\n")

    if len(sys.argv) > 1 and sys.argv[1] == "--batch":
        batch_mode()
        return

    if len(sys.argv) > 1 and sys.argv[1] == "--formats":
        url = sys.argv[2] if len(sys.argv) > 2 else input("🔗 URL: ").strip()
        list_formats(url)
        return

    url = input("🔗 Enter URL: ").strip()
    if not url:
        print("❌ No URL entered.")
        return

    ctype = detect_content_type(url)
    print(f"🧠 Detected: {ctype.upper()}")

    if ctype in ["video", "audio"]:
        download_yt_content(url, ctype)
    else:
        generic_file_download(url)

    compress = input("\n📦 Compress downloads folder into ZIP? (y/n): ").lower()
    if compress == "y":
        zip_folder("downloads")

    print("\n👋 Done!")


if __name__ == "__main__":
    main()
```

```python
"""
⚡ Universal Downloader X — Ultimate Edition v3.0
══════════════════════════════════════════════════
Features:
  ✅ Auto-resume on internet disconnect (keeps retrying until online)
  ✅ Pause / Resume with  P  key  (background keyboard listener)
  ✅ Multi-threaded parallel downloads (batch mode)
  ✅ Download history log (download_history.json)
  ✅ Speed limit control
  ✅ Proxy / VPN support
  ✅ Playlist & channel download
  ✅ Auto subtitle download (optional)
  ✅ Quality selection (360p → 4K)
  ✅ Audio-only MP3 with quality choice
  ✅ 403 / SABR fix via android client
  ✅ Batch mode from urls.txt
  ✅ ZIP compression after download
  ✅ Interactive main menu

Usage:
  python downloader.py                  → interactive menu
  python downloader.py --batch          → batch from urls.txt
  python downloader.py --history        → show download history
  python downloader.py --formats <url>  → list available formats
  python downloader.py <url>            → direct download
"""

import yt_dlp
import requests
import os
import sys
import time
import json
import threading
import socket
import zipfile
from pathlib import Path
from datetime import datetime
from concurrent.futures import ThreadPoolExecutor, as_completed

# ══════════════════════════════════════════════════
#  GLOBALS
# ══════════════════════════════════════════════════
HISTORY_FILE = "download_history.json"
paused       = False
pause_lock   = threading.Lock()


# ══════════════════════════════════════════════════
#  PAUSE / RESUME  (background keyboard listener)
# ══════════════════════════════════════════════════
def _toggle_pause():
    global paused
    with pause_lock:
        paused = not paused
    state = "⏸  PAUSED   — press P to resume" if paused else "▶️  RESUMED  — downloading..."
    print(f"\n{state}\n")


def _keyboard_listener():
    """Daemon thread: press P to pause/resume, Q to quit."""
    global paused
    try:
        if os.name == "nt":                    # Windows
            import msvcrt
            while True:
                if msvcrt.kbhit():
                    ch = msvcrt.getwch().lower()
                    if ch == "p":
                        _toggle_pause()
                    elif ch == "q":
                        print("\n👋  Quit requested.\n")
                        os._exit(0)
                time.sleep(0.2)
        else:                                   # Linux / macOS
            import tty, termios, select
            fd  = sys.stdin.fileno()
            old = termios.tcgetattr(fd)
            try:
                tty.setraw(fd)
                while True:
                    r, _, _ = select.select([sys.stdin], [], [], 0.2)
                    if r:
                        ch = sys.stdin.read(1).lower()
                        if ch == "p":
                            _toggle_pause()
                        elif ch == "q":
                            print("\n👋  Quit requested.\n")
                            os._exit(0)
            finally:
                termios.tcsetattr(fd, termios.TCSADRAIN, old)
    except Exception:
        pass  # silently ignore if stdin unavailable


def start_keyboard_listener():
    t = threading.Thread(target=_keyboard_listener, daemon=True)
    t.start()


def wait_if_paused():
    """Blocks execution while paused. Call inside download loops."""
    while True:
        with pause_lock:
            if not paused:
                break
        time.sleep(0.5)


# ══════════════════════════════════════════════════
#  INTERNET CHECK & AUTO-RECONNECT
# ══════════════════════════════════════════════════
def is_online(host="8.8.8.8", port=53, timeout=3):
    """Ping Google DNS to check connectivity."""
    try:
        socket.setdefaulttimeout(timeout)
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.connect((host, port))
        s.close()
        return True
    except OSError:
        return False


def wait_for_internet(retry_secs=5, max_wait=600):
    """
    Block until internet comes back.
    Shows live countdown. Returns True when online, False on timeout.
    """
    if is_online():
        return True
    print("\n🔌  Internet disconnected — waiting to reconnect...")
    elapsed = 0
    while elapsed < max_wait:
        for remaining in range(retry_secs, 0, -1):
            print(f"\r   Retrying in {remaining}s  (total waited: {elapsed}s)   ", end="", flush=True)
            time.sleep(1)
        elapsed += retry_secs
        if is_online():
            print("\n✅  Internet restored — resuming!\n")
            return True
    print(f"\n❌  No internet after {max_wait}s — giving up.")
    return False


# ══════════════════════════════════════════════════
#  DOWNLOAD HISTORY
# ══════════════════════════════════════════════════
def load_history():
    if os.path.exists(HISTORY_FILE):
        try:
            with open(HISTORY_FILE, "r", encoding="utf-8") as f:
                return json.load(f)
        except Exception:
            pass
    return []


def save_history(entry: dict):
    history = load_history()
    history.append(entry)
    with open(HISTORY_FILE, "w", encoding="utf-8") as f:
        json.dump(history, f, indent=2, ensure_ascii=False)


def show_history():
    history = load_history()
    if not history:
        print("\n📭  No download history yet.")
        return
    print(f"\n📜  Download History  ({len(history)} total — showing last 30)")
    print("─" * 70)
    for i, e in enumerate(reversed(history[-30:]), 1):
        icon  = "✅" if e.get("status") == "ok" else "❌"
        ts    = e.get("time", "")[:16]
        title = e.get("title", e.get("url", "?"))[:50]
        print(f"  {i:>3}. {icon}  [{ts}]  {title}")
    print("─" * 70)


# ══════════════════════════════════════════════════
#  UTILITY
# ══════════════════════════════════════════════════
def safe_filename(name):
    for ch in '\\/*?:"<>|':
        name = name.replace(ch, "_")
    return name.strip()


def make_folder(base, sub=None):
    now  = datetime.now()
    path = Path(base) / str(now.year) / f"{now.month:02d}"
    if sub:
        path /= sub
    path.mkdir(parents=True, exist_ok=True)
    return str(path)


def zip_folder(folder):
    zip_name = f"{folder.rstrip(os.sep)}.zip"
    with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zf:
        for root, _, files in os.walk(folder):
            for file in files:
                full = os.path.join(root, file)
                zf.write(full, os.path.relpath(full, folder))
    print(f"📦  Compressed → {zip_name}")


def detect_content_type(url):
    u = url.lower()
    if any(x in u for x in ["youtube", "youtu.be", "tiktok", "vimeo",
                              "facebook", "x.com", "twitter", "instagram", "reddit"]):
        return "video"
    if any(x in u for x in ["soundcloud", "bandcamp"]):
        return "audio"
    if u.endswith((".jpg", ".jpeg", ".png", ".gif", ".webp")):
        return "image"
    if u.endswith(".pdf"):
        return "pdf"
    if u.endswith((".zip", ".rar", ".7z")):
        return "archive"
    if u.endswith((".doc", ".docx", ".ppt", ".pptx", ".xls", ".xlsx")):
        return "document"
    return "unknown"


def smart_folder(content_type):
    if content_type in ("video", "playlist"):
        return make_folder("downloads", "videos")
    if content_type == "audio":
        return make_folder("downloads", "audio")
    return make_folder("downloads", "files")


# ══════════════════════════════════════════════════
#  MENUS
# ══════════════════════════════════════════════════
QUALITY_MAP = {
    "1": ("2160", "4K Ultra HD  (2160p)"),
    "2": ("1440", "2K / QHD     (1440p)"),
    "3": ("1080", "Full HD      (1080p)  ← Recommended"),
    "4": ("720",  "HD Ready     (720p)"),
    "5": ("480",  "SD           (480p)"),
    "6": ("360",  "Low          (360p)"),
    "7": ("best", "Best available (auto)"),
}


def ask_quality(default="3"):
    print("\n🎬  Video Quality:")
    print("  " + "─" * 36)
    for k, (_, label) in QUALITY_MAP.items():
        arrow = " ◀" if k == default else ""
        print(f"  [{k}] {label}{arrow}")
    print("  " + "─" * 36)
    choice = input(f"  Number (default={default}): ").strip() or default
    res, label = QUALITY_MAP.get(choice, QUALITY_MAP[default])
    print(f"  ✅  {label}")
    return res


def ask_audio_quality(default="2"):
    opts   = {"1": "320", "2": "192", "3": "128", "4": "64"}
    labels = {
        "1": "320 kbps  (Best)",
        "2": "192 kbps  ← Recommended",
        "3": "128 kbps",
        "4": "64 kbps   (Smallest)",
    }
    print("\n🎵  Audio Quality:")
    for k, v in labels.items():
        arrow = " ◀" if k == default else ""
        print(f"  [{k}] {v}{arrow}")
    choice = input(f"  Number (default={default}): ").strip() or default
    return opts.get(choice, "192")


def ask_speed_limit():
    print("\n🚀  Speed Limit  (blank = unlimited)  e.g. 500K / 2M / 1.5M:")
    val = input("  Limit: ").strip()
    return val or None


def ask_proxy():
    print("\n🌐  Proxy  (blank = none)")
    print("  e.g.  socks5://127.0.0.1:1080   or   http://user:pass@host:port")
    val = input("  Proxy: ").strip()
    return val or None


def ask_subtitles():
    return input("\n📝  Download subtitles? (y/n, default=n): ").strip().lower() == "y"


def ask_workers(max_n=8):
    val = input(f"\n⚡  Parallel workers (1–{max_n}, default=3): ").strip()
    try:
        return max(1, min(int(val), max_n))
    except ValueError:
        return 3


# ══════════════════════════════════════════════════
#  BUILD yt-dlp OPTIONS
# ══════════════════════════════════════════════════
def build_opts(folder, resolution="1080", audio_only=False,
               audio_quality="192", proxy=None, limit_rate=None,
               subtitles=False, playlist=True):

    if audio_only:
        fmt = "bestaudio/best"
    elif resolution == "best":
        fmt = "bestvideo+bestaudio/best"
    else:
        fmt = (
            f"bestvideo[height<={resolution}][ext=mp4]+bestaudio[ext=m4a]"
            f"/bestvideo[height<={resolution}]+bestaudio"
            f"/best[height<={resolution}]"
            f"/bestvideo+bestaudio/best"
        )

    opts = {
        "outtmpl":             os.path.join(folder, "%(title)s [%(height)sp].%(ext)s"),
        "merge_output_format": "mp4",
        "format":              fmt,

        # ── 403 / SABR fix ─────────────────────────────────────────
        "extractor_args":      {"youtube": {"player_client": ["android", "web"]}},

        # ── Resilience ─────────────────────────────────────────────
        "retries":             15,
        "fragment_retries":    15,
        "continuedl":          True,   # resume partial file
        "ignoreerrors":        True,
        "noplaylist":          not playlist,

        # ── Metadata ───────────────────────────────────────────────
        "writethumbnail":      True,
        "writeinfojson":       True,
        "write_description":   True,
        "writesubtitles":      subtitles,
        "writeautomaticsub":   subtitles,
        "subtitleslangs":      ["en", "hi"] if subtitles else [],

        # ── Output ─────────────────────────────────────────────────
        "quiet":               False,
        "progress":            True,

        # ── Post-processing ────────────────────────────────────────
        "postprocessors": [
            {"key": "FFmpegMetadata"},
            {"key": "EmbedThumbnail"},
        ],
    }

    if proxy:
        opts["proxy"] = proxy
    if limit_rate:
        opts["ratelimit"] = limit_rate

    if audio_only:
        opts["postprocessors"].append({
            "key":              "FFmpegExtractAudio",
            "preferredcodec":   "mp3",
            "preferredquality": audio_quality,
        })

    return opts


# ══════════════════════════════════════════════════
#  LIST FORMATS
# ══════════════════════════════════════════════════
def list_formats(url):
    print(f"\n📋  Fetching formats: {url}\n")
    try:
        with yt_dlp.YoutubeDL({
            "listformats": True,
            "quiet":       False,
            "extractor_args": {"youtube": {"player_client": ["android", "web"]}},
        }) as ydl:
            ydl.download([url])
    except Exception as e:
        print(f"⚠️  {e}")


# ══════════════════════════════════════════════════
#  CORE DOWNLOAD  (auto-reconnect + pause)
# ══════════════════════════════════════════════════

# Keywords that mean "network is gone" — not a format/auth error
_NET_ERRORS = [
    "network is unreachable", "name or service not known",
    "failed to resolve", "errno -2", "errno 101", "errno 110",
    "timed out", "read timed out", "connection reset",
    "connection refused", "broken pipe", "remote end closed",
    "10054", "10060", "10061", "nodename nor servname",
    "temporary failure in name resolution",
]

def _is_network_error(text: str) -> bool:
    t = text.lower()
    return any(kw in t for kw in _NET_ERRORS)


def _make_progress_hook():
    """
    Returns a yt-dlp progress hook that:
      1. Honours the global pause flag (blocks inside the hook)
      2. Detects a stalled/errored fragment and waits for internet
         BEFORE yt-dlp exhausts its own retry count.
    """
    last_bytes  = [0]
    last_time   = [time.time()]
    stall_limit = 30   # seconds with 0 new bytes → treat as disconnect

    def hook(d):
        wait_if_paused()

        if d["status"] == "downloading":
            downloaded = d.get("downloaded_bytes", 0)
            now        = time.time()

            if downloaded != last_bytes[0]:
                last_bytes[0] = downloaded
                last_time[0]  = now
            else:
                # No progress — check if stalled
                stalled_for = now - last_time[0]
                if stalled_for >= stall_limit and not is_online():
                    print(f"\n🔌  Stall detected ({stalled_for:.0f}s no data) — internet disconnected")
                    wait_for_internet(retry_secs=5)
                    last_time[0] = time.time()   # reset stall clock after reconnect

        elif d["status"] == "error":
            err = str(d.get("error", ""))
            if _is_network_error(err):
                print(f"\n🔌  Fragment error — waiting for internet...")
                wait_for_internet(retry_secs=5)

    return hook


def _do_download(url, opts, max_reconnect=999):
    """
    Wraps yt_dlp download with:
      • progress hook → pauses yt-dlp mid-download when internet drops
      • outer retry loop → restarts the entire download if yt-dlp gives up
      • continuedl=True  → always resumes the partial .mp4/.part file
      • unlimited reconnect attempts (max_reconnect=999 by default)

    This means:
      - Internet drops mid-download  → hook detects it, waits, yt-dlp resumes
      - yt-dlp exhausts its retries  → outer loop waits for internet, restarts
      - Partial file is NEVER deleted → always continues from last byte

    Returns ('ok', title) or ('fail', error_msg)
    """
    title = url

    # Inject our progress hook (keep any existing hooks the caller added)
    existing_hooks = opts.get("progress_hooks", [])
    opts["progress_hooks"] = existing_hooks + [_make_progress_hook()]

    # Force yt-dlp retries high so the hook has time to wait for reconnect
    opts["retries"]          = 99
    opts["fragment_retries"] = 99
    opts["continuedl"]       = True   # NEVER restart from zero

    attempt = 0
    while attempt <= max_reconnect:
        wait_if_paused()

        # Hard-check before starting/restarting
        if not is_online():
            print(f"\n🔌  No internet — waiting before attempt {attempt + 1}...")
            if not wait_for_internet():
                return ("fail", "No internet after maximum wait time")

        try:
            with yt_dlp.YoutubeDL(opts) as ydl:
                info = ydl.extract_info(url, download=True)
                if info:
                    title = info.get("title", url)
            return ("ok", title)   # ← success

        except yt_dlp.utils.DownloadError as e:
            err = str(e)
            if _is_network_error(err):
                attempt += 1
                print(f"\n🔌  yt-dlp gave up after retries (disconnect #{attempt})")
                print(   "    Waiting for internet then resuming partial file...\n")
                wait_for_internet(retry_secs=8)
                # Loop back — continuedl=True means it picks up the .part file
                continue
            # Non-network error (e.g. geo-block, bad format) — don't retry
            return ("fail", err)

        except Exception as e:
            err = str(e)
            if _is_network_error(err):
                attempt += 1
                wait_for_internet(retry_secs=8)
                continue
            return ("fail", err)

    return ("fail", f"Gave up after {max_reconnect} disconnect recoveries")


# ══════════════════════════════════════════════════
#  SINGLE VIDEO / AUDIO
# ══════════════════════════════════════════════════
def download_yt_content(url, content_type, proxy=None, limit_rate=None):
    folder    = smart_folder(content_type)
    subtitles = ask_subtitles()

    if content_type == "audio":
        aq   = ask_audio_quality()
        opts = build_opts(folder, audio_only=True, audio_quality=aq,
                          proxy=proxy, limit_rate=limit_rate, subtitles=subtitles)
        res  = "audio"
    else:
        if input("\n👀  Show available formats first? (y/n): ").strip().lower() == "y":
            list_formats(url)
        res  = ask_quality()
        aq   = ask_audio_quality()
        opts = build_opts(folder, resolution=res, audio_quality=aq,
                          proxy=proxy, limit_rate=limit_rate, subtitles=subtitles)

    print(f"\n⏬  {content_type.upper()}  |  quality: {res}  |  → {folder}")
    print(   "   Press  P  to pause/resume   |   Q  to quit\n")

    status, result = _do_download(url, opts)

    save_history({
        "time":       datetime.now().isoformat(timespec="seconds"),
        "url":        url,
        "title":      result if status == "ok" else url,
        "type":       content_type,
        "resolution": res,
        "status":     status,
        "folder":     folder,
    })

    if status == "ok":
        print(f"\n✅  Done!  Saved → {folder}")
    else:
        print(f"\n❌  Failed: {result}")
        print("💡  pip install -U yt-dlp  then try again")


# ══════════════════════════════════════════════════
#  GENERIC FILE  (PDF / image / doc)
# ══════════════════════════════════════════════════
def generic_file_download(url, max_retries=5):
    folder  = smart_folder("files")
    headers = {"User-Agent": "Mozilla/5.0"}
    print(f"\n📥  {url}")

    for attempt in range(1, max_retries + 1):
        wait_if_paused()
        if not is_online():
            wait_for_internet()
        try:
            r     = requests.get(url, stream=True, timeout=30, headers=headers)
            if r.status_code != 200:
                print(f"❌  HTTP {r.status_code}")
                return
            ctype    = r.headers.get("content-type", "")
            ext      = ctype.split("/")[-1].split(";")[0] if "/" in ctype else "bin"
            fname    = safe_filename(url.split("/")[-1].split("?")[0]) or f"file_{int(time.time())}.{ext}"
            filepath = os.path.join(folder, fname)
            total    = int(r.headers.get("content-length", 0))
            done     = 0
            start    = time.time()
            with open(filepath, "ab") as f:
                for chunk in r.iter_content(65536):
                    wait_if_paused()
                    f.write(chunk)
                    done += len(chunk)
                    if total:
                        pct     = done / total * 100
                        elapsed = max(time.time() - start, 0.001)
                        speed   = done / elapsed / 1024
                        print(f"\r  {pct:5.1f}%  {done//1024}KB/{total//1024}KB  {speed:.0f}KB/s  ", end="", flush=True)
            print(f"\n✅  Saved → {filepath}")
            save_history({"time": datetime.now().isoformat(timespec="seconds"),
                          "url": url, "status": "ok", "folder": folder})
            return
        except requests.exceptions.ConnectionError:
            print(f"\n🔌  Connection lost (attempt {attempt}/{max_retries})")
            wait_for_internet()
        except Exception as e:
            print(f"\n⚠️  {e}")
            return
    print("❌  Max retries reached.")


# ══════════════════════════════════════════════════
#  BATCH MODE  (parallel)
# ══════════════════════════════════════════════════
def _batch_worker(args):
    url, opts, ctype = args
    if ctype in ("video", "audio"):
        return (url, *_do_download(url, opts))
    else:
        generic_file_download(url)
        return (url, "ok", url)


def batch_mode(file_path="urls.txt"):
    if not os.path.exists(file_path):
        print(f"⚠️  File not found: {file_path}")
        print("    Create it with one URL per line. Lines starting with # are skipped.")
        return

    with open(file_path, "r", encoding="utf-8") as f:
        urls = [l.strip() for l in f if l.strip() and not l.startswith("#")]

    if not urls:
        print("⚠️  No URLs found.")
        return

    print(f"\n📦  Batch mode — {len(urls)} URLs")
    res        = ask_quality()
    aq         = ask_audio_quality()
    subtitles  = ask_subtitles()
    proxy      = ask_proxy()
    limit_rate = ask_speed_limit()
    workers    = ask_workers()

    tasks = []
    for url in urls:
        ctype  = detect_content_type(url)
        folder = smart_folder(ctype)
        opts   = build_opts(folder, resolution=res, audio_only=(ctype == "audio"),
                            audio_quality=aq, proxy=proxy, limit_rate=limit_rate,
                            subtitles=subtitles)
        tasks.append((url, opts, ctype))

    print(f"\n⚡  Starting {workers} parallel workers...")
    print("   Press  P  to pause/resume  |  Q  to quit\n")

    ok_count = fail_count = 0
    with ThreadPoolExecutor(max_workers=workers) as pool:
        futures = {pool.submit(_batch_worker, t): t[0] for t in tasks}
        for future in as_completed(futures):
            url = futures[future]
            try:
                _, status, info = future.result()
                if status == "ok":
                    ok_count += 1
                    print(f"  ✅  {str(info)[:60]}")
                else:
                    fail_count += 1
                    print(f"  ❌  {url[:50]}  →  {str(info)[:40]}")
                save_history({"time": datetime.now().isoformat(timespec="seconds"),
                              "url": url, "title": str(info), "status": status})
            except Exception as e:
                fail_count += 1
                print(f"  ⚠️  {url[:50]}  →  {e}")

    print(f"\n═══ Batch done ═══  ✅ {ok_count} succeeded   ❌ {fail_count} failed")
    if input("\n📦  Compress all downloads to ZIP? (y/n): ").strip().lower() == "y":
        zip_folder("downloads")


# ══════════════════════════════════════════════════
#  INTERACTIVE MENU
# ══════════════════════════════════════════════════
def interactive_menu():
    while True:
        print("\n╔═══════════════════════════════════════════════╗")
        print("║   ⚡  Universal Downloader X  v3.0            ║")
        print("╠═══════════════════════════════════════════════╣")
        print("║  [1]  Download video / playlist               ║")
        print("║  [2]  Download audio only (MP3)               ║")
        print("║  [3]  Download file (PDF / image / doc)       ║")
        print("║  [4]  Batch download  (urls.txt)              ║")
        print("║  [5]  Show available formats for a URL        ║")
        print("║  [6]  View download history                   ║")
        print("║  [7]  Compress downloads folder to ZIP        ║")
        print("║  [0]  Exit                                    ║")
        print("╚═══════════════════════════════════════════════╝")
        choice = input("  Choose: ").strip()

        if choice == "0":
            print("👋  Bye!")
            sys.exit(0)

        elif choice in ("1", "2"):
            url = input("🔗  URL: ").strip()
            if not url:
                continue
            proxy      = ask_proxy()
            limit_rate = ask_speed_limit()
            ctype      = "video" if choice == "1" else "audio"
            download_yt_content(url, ctype, proxy=proxy, limit_rate=limit_rate)

        elif choice == "3":
            url = input("🔗  File URL: ").strip()
            if url:
                generic_file_download(url)

        elif choice == "4":
            path = input("📄  URL list path (default=urls.txt): ").strip() or "urls.txt"
            batch_mode(path)

        elif choice == "5":
            url = input("🔗  URL: ").strip()
            if url:
                list_formats(url)

        elif choice == "6":
            show_history()

        elif choice == "7":
            folder = input("📁  Folder (default=downloads): ").strip() or "downloads"
            if os.path.isdir(folder):
                zip_folder(folder)
            else:
                print(f"⚠️  Not found: {folder}")
        else:
            print("⚠️  Invalid choice — try again.")


# ══════════════════════════════════════════════════
#  ENTRY POINT
# ══════════════════════════════════════════════════
def main():
    start_keyboard_listener()   # P = pause/resume, Q = quit

    if len(sys.argv) > 1:
        cmd = sys.argv[1]
        if cmd == "--batch":
            batch_mode(sys.argv[2] if len(sys.argv) > 2 else "urls.txt")
        elif cmd == "--history":
            show_history()
        elif cmd == "--formats":
            url = sys.argv[2] if len(sys.argv) > 2 else input("🔗  URL: ").strip()
            list_formats(url)
        elif cmd == "--help":
            print(__doc__)
        else:
            # Treat first arg as a direct URL
            url   = cmd
            ctype = detect_content_type(url)
            print(f"🧠  Detected: {ctype.upper()}")
            download_yt_content(url, ctype) if ctype in ("video", "audio") else generic_file_download(url)
    else:
        interactive_menu()


if __name__ == "__main__":
    main()
```
