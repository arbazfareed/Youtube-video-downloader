```python
#!/usr/bin/env python3
"""
==============================================
⚡ SMART DOWNLOADER v9 — Season/Show Edition
==============================================
Features:
- Full playlist / single video download
- Season pack support (TV shows)
- List ALL available video qualities before downloading
- Smart quality fallback (1080p → 720p → 480p)
- Audio MP3 extraction
- Cookie fallback (Chrome → no-cookie public mode)
- Resume broken downloads
- Skip already-downloaded files
- Conda-compatible (no broken pip issues)
==============================================
"""

import os
import sys
import yt_dlp
from datetime import datetime


# ─────────────────────────────────────────────
# VALIDATION
# ─────────────────────────────────────────────

def is_valid_url(url: str) -> bool:
    return url.startswith("http://") or url.startswith("https://")

def is_playlist(url: str) -> bool:
    return "playlist?list=" in url or "&list=" in url or "/playlist/" in url


# ─────────────────────────────────────────────
# FOLDER SYSTEM
# ─────────────────────────────────────────────

def get_folder(custom_name: str = None) -> str:
    base = "downloads"
    date_folder = datetime.now().strftime("%Y-%m")
    if custom_name:
        # Sanitize folder name
        safe_name = "".join(c if c.isalnum() or c in " _-" else "_" for c in custom_name).strip()
        path = os.path.join(base, date_folder, safe_name)
    else:
        path = os.path.join(base, date_folder)
    os.makedirs(path, exist_ok=True)
    return path


# ─────────────────────────────────────────────
# LIST ALL AVAILABLE FORMATS (with file sizes)
# ─────────────────────────────────────────────

def list_available_formats(url: str):
    """
    Shows all available video resolutions and their estimated file sizes.
    Works for single videos and the first video in a playlist.
    """
    print("\n🔍 Fetching available formats...\n")

    ydl_opts = {
        "quiet": True,
        "no_warnings": True,
        "skip_download": True,
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=False)

            # If playlist, grab first entry
            if "entries" in info:
                print(f"📋 Playlist detected: {info.get('title', 'Unknown')}")
                print("   Showing formats for the FIRST video in the playlist.\n")
                info = next(iter(info["entries"]), None)
                if not info:
                    print("❌ Could not extract first video from playlist.")
                    return

            title = info.get("title", "Unknown")
            print(f"🎬 Video: {title}\n")
            print(f"{'ID':<12} {'EXT':<6} {'RESOLUTION':<12} {'FPS':<6} {'SIZE':<12} {'NOTE'}")
            print("─" * 70)

            formats = info.get("formats", [])
            seen = set()

            for f in formats:
                fmt_id    = f.get("format_id", "?")
                ext       = f.get("ext", "?")
                height    = f.get("height")
                fps       = f.get("fps") or "-"
                filesize  = f.get("filesize") or f.get("filesize_approx")
                note      = f.get("format_note", "")
                vcodec    = f.get("vcodec", "none")
                acodec    = f.get("acodec", "none")

                # Only show video formats (skip audio-only lines here)
                if vcodec == "none":
                    continue

                res = f"{height}p" if height else "audio"
                key = (res, ext)
                if key in seen:
                    continue
                seen.add(key)

                if filesize:
                    size_mb = f"{filesize / (1024*1024):.1f} MB"
                else:
                    size_mb = "~unknown"

                print(f"{fmt_id:<12} {ext:<6} {res:<12} {str(fps):<6} {size_mb:<12} {note}")

            # Also list audio-only
            print("\n── Audio Only ──")
            print(f"{'ID':<12} {'EXT':<6} {'BITRATE':<12} {'SIZE':<12} {'NOTE'}")
            print("─" * 55)
            for f in formats:
                vcodec = f.get("vcodec", "none")
                acodec = f.get("acodec", "none")
                if vcodec != "none":
                    continue
                fmt_id   = f.get("format_id", "?")
                ext      = f.get("ext", "?")
                abr      = f.get("abr") or "-"
                filesize = f.get("filesize") or f.get("filesize_approx")
                note     = f.get("format_note", "")
                size_mb  = f"{filesize/(1024*1024):.1f} MB" if filesize else "~unknown"
                print(f"{fmt_id:<12} {ext:<6} {str(abr)+' kbps':<12} {size_mb:<12} {note}")

            print(f"\n💡 Tip: Use format ID directly → enter it as custom quality in Download menu.")

    except Exception as e:
        print(f"❌ Could not fetch formats: {e}")


# ─────────────────────────────────────────────
# SMART QUALITY SELECTOR
# ─────────────────────────────────────────────

def smart_format(q: str) -> str:
    """
    q can be:
      1  = 4K
      2  = 1440p
      3  = 1080p (smart fallback: tries 1080 → 720 → 480 → best)
      4  = 720p
      5  = 480p
      6  = 360p
      7  = best available
      OR a raw yt-dlp format ID (e.g. '137+140')
    """
    quality_map = {
        "1": "bestvideo[height<=2160]+bestaudio/bestvideo+bestaudio/best",
        "2": "bestvideo[height<=1440]+bestaudio/bestvideo+bestaudio/best",
        "3": (
            "bestvideo[height<=1080]+bestaudio/"
            "bestvideo[height<=720]+bestaudio/"
            "bestvideo[height<=480]+bestaudio/"
            "best"
        ),
        "4": "bestvideo[height<=720]+bestaudio/best",
        "5": "bestvideo[height<=480]+bestaudio/best",
        "6": "bestvideo[height<=360]+bestaudio/best",
        "7": "bestvideo+bestaudio/best",
    }
    return quality_map.get(q, q)  # if unknown, treat as raw format string


# ─────────────────────────────────────────────
# MAIN DOWNLOAD ENGINE
# ─────────────────────────────────────────────

def download(url: str, mode: str = "video", quality: str = "3", show_name: str = None):
    if not is_valid_url(url):
        print("❌ Invalid URL. Make sure it starts with http:// or https://")
        return

    folder = get_folder(show_name)
    archive_file = os.path.join(folder, "downloaded.txt")

    print(f"\n📁 Save folder: {folder}")
    print(f"📦 Archive    : {archive_file}\n")

    # ── Output template ──
    # Season/episode-aware naming if playlist
    out_template = os.path.join(
        folder,
        "%(playlist_title)s",
        "%(playlist_index)s - %(title)s.%(ext)s"
    )
    # For single videos, playlist_title will just be blank or equal to title
    # yt-dlp handles this gracefully

    base_opts = {
        "outtmpl"            : out_template,
        "ignoreerrors"       : True,       # skip broken/geo-blocked videos
        "continuedl"         : True,       # resume incomplete downloads
        "retries"            : 20,
        "fragment_retries"   : 20,
        "nooverwrites"       : True,
        "download_archive"   : archive_file,
        "merge_output_format": "mp4",
        "writesubtitles"     : False,      # set True if you want subtitles
        "extractor_args"     : {
            "youtube": {
                "player_client": ["android", "web"]
            }
        },
        "http_headers": {
            "User-Agent": (
                "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                "AppleWebKit/537.36 (KHTML, like Gecko) "
                "Chrome/120.0.0.0 Safari/537.36"
            )
        },
    }

    # ── Run function ──
    def run(use_cookies: bool = True):
        opts = base_opts.copy()

        if use_cookies:
            opts["cookiesfrombrowser"] = ("chrome",)

        if mode == "audio":
            opts["format"] = "bestaudio/best"
            opts["postprocessors"] = [{
                "key"             : "FFmpegExtractAudio",
                "preferredcodec"  : "mp3",
                "preferredquality": "192",
            }]
        else:
            opts["format"] = smart_format(quality)

        with yt_dlp.YoutubeDL(opts) as ydl:
            return ydl.download([url])

    # ── Execution with cookie fallback ──
    try:
        print("⏬ Attempting download with Chrome cookies...\n")
        run(use_cookies=True)

    except Exception as e:
        print(f"\n⚠️  Chrome cookies failed → switching to public mode...\n   Reason: {e}\n")
        try:
            print("⏬ Retrying WITHOUT cookies (public mode)...\n")
            run(use_cookies=False)

        except Exception as e2:
            print(f"\n❌ COMPLETE FAILURE: {e2}")
            return

    print("\n✅ DONE!")
    print(f"📁 Files saved in: {folder}")


# ─────────────────────────────────────────────
# FIND SEASON PLAYLIST LINKS (helper guide)
# ─────────────────────────────────────────────

def show_how_to_find_links():
    guide = """
╔══════════════════════════════════════════════════════════════╗
║         HOW TO FIND SEASON / SHOW DOWNLOAD LINKS            ║
╠══════════════════════════════════════════════════════════════╣
║                                                              ║
║  yt-dlp supports 1000+ sites. For TV shows/seasons:         ║
║                                                              ║
║  ① YouTube                                                   ║
║     Search: "Show Name Season 1 full playlist"               ║
║     URL looks like:                                          ║
║     https://www.youtube.com/playlist?list=PLxxxxxxx          ║
║     → Copy the full URL from the browser address bar         ║
║                                                              ║
║  ② Dailymotion                                               ║
║     https://www.dailymotion.com/playlist/xxxxx               ║
║                                                              ║
║  ③ Archive.org (free, legal classics)                        ║
║     https://archive.org/details/show-name                    ║
║                                                              ║
║  ④ Vimeo (paid/free shows)                                   ║
║     https://vimeo.com/channels/channelname                   ║
║                                                              ║
║  ⑤ Any direct video URL                                      ║
║     Just paste the episode URL for a single download         ║
║                                                              ║
║  HOW TO GET THE LINK:                                        ║
║  1. Open the playlist/show page in your browser              ║
║  2. Copy the URL from the address bar                        ║
║  3. Paste it into this downloader                            ║
║                                                              ║
║  TIP: Use option [4] List Formats FIRST to see all           ║
║       available resolutions before downloading.              ║
║                                                              ║
╚══════════════════════════════════════════════════════════════╝
"""
    print(guide)


# ─────────────────────────────────────────────
# MENU
# ─────────────────────────────────────────────

def main():
    while True:
        print("""
╔══════════════════════════════════════╗
║      ⚡ SMART DOWNLOADER v9         ║
║   Season / Show / Playlist Edition  ║
╠══════════════════════════════════════╣
║  [1] Download Video / Playlist       ║
║  [2] Download Audio (MP3)            ║
║  [3] List ALL Available Formats      ║
║  [4] How to Find Season Links        ║
║  [0] Exit                            ║
╚══════════════════════════════════════╝""")

        choice = input("\nSelect: ").strip()

        if choice == "0":
            print("👋 Bye!")
            break

        elif choice == "4":
            show_how_to_find_links()
            continue

        # For options 1-3, we need a URL
        url = input("\nPaste URL (video, playlist, or season): ").strip()
        if not url:
            print("❌ No URL entered.")
            continue
        if not is_valid_url(url):
            print("❌ Invalid URL. Must start with http:// or https://")
            continue

        if choice == "1":
            print("""
┌─────────────────────────────────┐
│         Quality Options         │
├──────┬──────────────────────────┤
│  1   │ 4K (2160p)               │
│  2   │ 1440p                    │
│  3   │ 1080p ← SMART fallback   │
│  4   │ 720p                     │
│  5   │ 480p                     │
│  6   │ 360p                     │
│  7   │ Best available           │
│  ID  │ Custom format ID         │
└──────┴──────────────────────────┘
Tip: Run [3] first to see exact format IDs & file sizes.""")

            q = input("\nChoose quality [default=3]: ").strip() or "3"

            show_name = input("Show/Season name for folder (optional, press Enter to skip): ").strip()
            download(url, mode="video", quality=q, show_name=show_name or None)

        elif choice == "2":
            show_name = input("Show/Album name for folder (optional): ").strip()
            download(url, mode="audio", show_name=show_name or None)

        elif choice == "3":
            list_available_formats(url)

        else:
            print("❌ Invalid choice. Enter 0-4.")


# ─────────────────────────────────────────────
# ENTRY POINT
# ─────────────────────────────────────────────

if __name__ == "__main__":
    # Quick dependency check
    try:
        import yt_dlp
    except ImportError:
        print("❌ yt-dlp not installed!")
        print("   Run: conda install -c conda-forge yt-dlp")
        print("   Or:  pip install yt-dlp --break-system-packages")
        sys.exit(1)

    main()
```
