```python
### !/usr/bin/env python3
import os
import yt_dlp
from datetime import datetime

# -----------------------------
# URL VALIDATION
# -----------------------------

def is_valid_url(url: str) -> bool:
    return url.startswith("http")

def is_playlist(url: str) -> bool:
    return "playlist?list=" in url or "&list=" in url

# -----------------------------
# FOLDER SYSTEM
# -----------------------------

def get_folder():
    path = f"downloads/{datetime.now().strftime('%Y-%m')}"
    os.makedirs(path, exist_ok=True)
    return path

# -----------------------------
# QUALITY OPTIONS
# -----------------------------

def get_format(q="3"):
    return {
        "1": "bv*+ba/b[height<=2160]",
        "2": "bv*+ba/b[height<=1440]",
        "3": "bv*+ba/b[height<=1080]",
        "4": "bv*+ba/b[height<=720]",
        "5": "bestaudio/best"
    }.get(q, "bv*+ba/b[height<=1080]")

# -----------------------------
# CORE DOWNLOAD FUNCTION
# -----------------------------

def download(url: str, mode="video", quality="3"):
    if not is_valid_url(url):
        print("❌ Invalid URL")
        return

    folder = get_folder()

    # -------------------------
    # INNER RUN FUNCTION
    # -------------------------
    def run(use_browser_cookies=True):

        ydl_opts = {
            "outtmpl": f"{folder}/%(playlist_title)s/%(playlist_index)s - %(title)s.%(ext)s",

            # 🔥 STABILITY OPTIONS
            "ignoreerrors": True,
            "continuedl": True,
            "nooverwrites": True,
            "retries": 20,
            "fragment_retries": 20,

            # 🔥 IMPORTANT FOR PLAYLISTS
            "noplaylist": False,
            "merge_output_format": "mp4",

            # 🔥 AVOID DUPLICATES
            "download_archive": "downloaded.txt",
        }

        # -------------------------
        # COOKIE HANDLING
        # -------------------------
        if use_browser_cookies:
            ydl_opts["cookiesfrombrowser"] = ("chrome",)

        # -------------------------
        # VIDEO / AUDIO MODE
        # -------------------------
        if mode == "audio":
            ydl_opts["format"] = "bestaudio/best"
            ydl_opts["postprocessors"] = [{
                "key": "FFmpegExtractAudio",
                "preferredcodec": "mp3",
                "preferredquality": "192",
            }]
        else:
            ydl_opts["format"] = get_format(quality)

        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            return ydl.extract_info(url, download=True)

    # -----------------------------
    # MAIN EXECUTION
    # -----------------------------

    try:
        print("\n⏬ Starting download (Chrome cookies)...\n")
        info = run(True)

    except Exception as e:
        print(f"⚠ Chrome cookies failed: {e}")
        print("🔁 Trying fallback mode (no login)...")

        try:
            info = run(False)

        except Exception as e2:
            print(f"❌ Complete failure: {e2}")
            return

    print("\n✅ Download finished (partial or full supported)")
    print(f"📁 Saved in: {folder}")

# -----------------------------
# FORMAT LIST (optional)
# -----------------------------

def list_formats(url):
    with yt_dlp.YoutubeDL({"listformats": True}) as ydl:
        ydl.download([url])

# -----------------------------
# MENU SYSTEM
# -----------------------------

def main():
    while True:
        print("""
=========================
⚡ Downloader v7 PRO FIX
=========================
[1] Video / Playlist Download
[2] Audio Download (MP3)
[3] List Formats
[0] Exit
""")

        choice = input("Select: ").strip()

        if choice == "0":
            break

        url = input("Paste URL: ").strip()

        if choice == "1":
            q = input("Quality (1-4K / 2-1440 / 3-1080 / 4-720): ").strip()
            download(url, "video", q)

        elif choice == "2":
            download(url, "audio")

        elif choice == "3":
            list_formats(url)

        else:
            print("❌ Invalid choice")

# -----------------------------
# RUN PROGRAM
# -----------------------------

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
import os
import yt_dlp
from datetime import datetime

# -----------------------------
# URL CHECK
# -----------------------------

def is_valid_url(url: str) -> bool:
    return url.startswith("http")

# -----------------------------
# FOLDER SYSTEM
# -----------------------------

def get_folder():
    path = f"downloads/{datetime.now().strftime('%Y-%m')}"
    os.makedirs(path, exist_ok=True)
    return path

# -----------------------------
# SMART QUALITY FALLBACK SYSTEM
# -----------------------------

def get_smart_format(quality):
    """
    SMART FALLBACK ORDER:
    Try requested quality → then lower → then best available
    """

    if quality == "1":  # 4K
        return "bv*[height<=2160]+ba/b/bv*+ba/b[height<=1440]/bv*+ba/b[height<=1080]/best"

    if quality == "2":  # 1440p
        return "bv*[height<=1440]+ba/b/bv*+ba/b[height<=1080]/bv*+ba/b[height<=720]/best"

    if quality == "3":  # 1080p
        return "bv*[height<=1080]+ba/b/bv*+ba/b[height<=720]/bv*+ba/b[height<=480]/best"

    if quality == "4":  # 720p
        return "bv*[height<=720]+ba/b/bv*+ba/b[height<=480]/best"

    return "bv*+ba/best"

# -----------------------------
# CORE DOWNLOAD ENGINE
# -----------------------------

def download(url: str, mode="video", quality="3"):

    if not is_valid_url(url):
        print("❌ Invalid URL")
        return

    folder = get_folder()

    ydl_opts = {
        "outtmpl": f"{folder}/%(playlist_title)s/%(playlist_index)s - %(title)s.%(ext)s",

        # 🔥 STABILITY (VERY IMPORTANT)
        "ignoreerrors": True,
        "continuedl": True,
        "nooverwrites": True,
        "retries": 30,
        "fragment_retries": 30,

        # 🔥 NO DUPLICATES
        "download_archive": "downloaded.txt",

        # 🔥 AUTO FIX FORMAT ISSUES
        "format": get_smart_format(quality),

        # 🔥 FORCE MERGE
        "merge_output_format": "mp4",

        # 🔥 BYPASS SOME YOUTUBE LIMITS
        "nocheckcertificate": True,

        # 🔥 IMPORTANT FOR PLAYLISTS
        "noplaylist": False,
    }

    # -----------------------------
    # AUDIO MODE
    # -----------------------------
    if mode == "audio":
        ydl_opts["format"] = "bestaudio/best"
        ydl_opts["postprocessors"] = [{
            "key": "FFmpegExtractAudio",
            "preferredcodec": "mp3",
            "preferredquality": "192",
        }]

    try:
        print("\n⏬ Starting smart download...\n")

        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])

        print("\n✅ Download completed successfully")
        print(f"📁 Saved in: {folder}")

    except Exception as e:
        print(f"❌ Error: {e}")

# -----------------------------
# SIMPLE MENU
# -----------------------------

def main():
    while True:
        print("""
=========================
⚡ SMART Downloader v8
=========================
[1] Video / Playlist Download
[2] Audio Download (MP3)
[0] Exit
""")

        choice = input("Select: ").strip()

        if choice == "0":
            break

        url = input("Paste URL: ").strip()

        if choice == "1":
            print("""
Quality:
1 = 4K
2 = 1440p
3 = 1080p
4 = 720p
""")
            q = input("Choose quality: ").strip()
            download(url, "video", q)

        elif choice == "2":
            download(url, "audio")

        else:
            print("❌ Invalid choice")

# -----------------------------
# RUN
# -----------------------------

if __name__ == "__main__":
    main()
```
