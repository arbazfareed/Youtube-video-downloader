```python
import yt_dlp
import requests
import os
import sys
import time
from pathlib import Path
from datetime import datetime
import zipfile

# ───────────────────────────────
#  Utility Functions
# ───────────────────────────────
def safe_filename(name):
    """Clean up filenames for OS compatibility."""
    for char in '\\/*?:"<>|':
        name = name.replace(char, "_")
    return name.strip()


def make_folder(base, sub=None):
    """Create year/month structured folder."""
    now = datetime.now()
    path = Path(base) / str(now.year) / f"{now.month:02d}"
    if sub:
        path /= sub
    path.mkdir(parents=True, exist_ok=True)
    return str(path)


def zip_folder(folder):
    """Compress the downloaded folder into a zip archive."""
    zip_name = f"{folder.rstrip(os.sep)}.zip"
    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(folder):
            for file in files:
                full = os.path.join(root, file)
                zipf.write(full, os.path.relpath(full, folder))
    print(f"📦 Downloads compressed → {zip_name}")


def detect_content_type(url):
    """Guess content type from domain or extension."""
    url_l = url.lower()
    if any(x in url_l for x in ["youtube", "youtu.be", "tiktok", "vimeo", "facebook", "x.com", "twitter", "instagram", "reddit"]):
        return "video"
    if any(x in url_l for x in ["soundcloud", "bandcamp", "spotify"]):
        return "audio"
    if url_l.endswith((".jpg", ".jpeg", ".png", ".gif", ".webp")):
        return "image"
    if url_l.endswith(".pdf"):
        return "pdf"
    if url_l.endswith((".zip", ".rar", ".7z")):
        return "archive"
    if url_l.endswith((".doc", ".docx", ".ppt", ".pptx", ".xls", ".xlsx")):
        return "document"
    return "unknown"


def smart_download_folder(content_type):
    """Assign base folder based on type."""
    if content_type in ["video", "playlist"]:
        return make_folder("downloads", "videos")
    if content_type == "audio":
        return make_folder("downloads", "audio")
    return make_folder("downloads", "files")


# ───────────────────────────────
#  Generic File Downloader
# ───────────────────────────────
def generic_file_download(url):
    """Handles PDF, image, doc, etc."""
    folder = smart_download_folder("files")
    print(f"📥 Downloading generic file from: {url}")
    try:
        r = requests.get(url, stream=True, timeout=20)
        if r.status_code != 200:
            print("❌ Failed to fetch URL.")
            return
        content_type = r.headers.get("content-type", "")
        ext = content_type.split('/')[-1] if '/' in content_type else 'bin'
        filename = safe_filename(url.split('/')[-1].split('?')[0]) or f"file_{int(time.time())}.{ext}"
        filepath = os.path.join(folder, filename)
        with open(filepath, "wb") as f:
            for chunk in r.iter_content(8192):
                f.write(chunk)
        print(f"✅ Saved → {filepath}")
    except Exception as e:
        print(f"⚠️ Error downloading file: {e}")


# ───────────────────────────────
#  yt-dlp Video/Audio Logic
# ───────────────────────────────
def build_opts(folder, audio_only=False, proxy=None, limit_rate=None):
    opts = {
        "outtmpl": os.path.join(folder, "%(title)s.%(ext)s"),
        "noplaylist": False,
        "writethumbnail": True,
        "writesubtitles": True,
        "writeinfojson": True,
        "write_description": True,
        "merge_output_format": "mp4",
        "continuedl": True,
        "retries": 5,
        "ignoreerrors": True,
        "quiet": False,
        "progress": True,
        "postprocessors": [{"key": "FFmpegMetadata"}],
    }

    if proxy:
        opts["proxy"] = proxy
    if limit_rate:
        opts["ratelimit"] = limit_rate

    if audio_only:
        opts["format"] = "bestaudio/best"
        opts["postprocessors"].append({
            "key": "FFmpegExtractAudio",
            "preferredcodec": "mp3",
            "preferredquality": "192",
        })
        opts["postprocessors"].append({
            "key": "FFmpegAudioVolumeNormalize",
            "level": 1.0
        })
    else:
        opts["format"] = "bestvideo+bestaudio/best"

    return opts


def download_yt_content(url, content_type):
    """Main yt-dlp download logic."""
    folder = smart_download_folder(content_type)
    opts = build_opts(
        folder,
        audio_only=(content_type == "audio"),
        proxy=None,
        limit_rate=None
    )

    try:
        print(f"\n⏬ Downloading {content_type} from: {url}\n")
        with yt_dlp.YoutubeDL(opts) as ydl:
            ydl.download([url])
        print(f"✅ {content_type.capitalize()} saved in {folder}")
    except Exception as e:
        print(f"❌ Error downloading: {e}")


# ───────────────────────────────
#  Batch Mode
# ───────────────────────────────
def batch_mode(file_path="urls.txt"):
    """Download all URLs from urls.txt."""
    if not os.path.exists(file_path):
        print("⚠️ No urls.txt found for batch mode.")
        return
    with open(file_path, "r") as f:
        urls = [line.strip() for line in f if line.strip()]
    print(f"📦 Starting batch mode — {len(urls)} URLs found.")
    for url in urls:
        content_type = detect_content_type(url)
        if content_type in ["video", "audio"]:
            download_yt_content(url, content_type)
        else:
            generic_file_download(url)
    zip_folder("downloads")
    print("✅ Batch mode complete.")


# ───────────────────────────────
#  Main Entry
# ───────────────────────────────
def main():
    print("⚡ Universal Downloader X — Ultimate Edition")
    print("─────────────────────────────────────────────\n")

    # Batch mode
    if len(sys.argv) > 1 and sys.argv[1] == "--batch":
        batch_mode()
        return

    url = input("🔗 Enter any URL: ").strip()
    if not url:
        print("❌ No URL entered.")
        return

    ctype = detect_content_type(url)
    print(f"🧠 Detected content type: {ctype.upper()}")

    if ctype in ["video", "audio"]:
        download_yt_content(url, ctype)
    else:
        generic_file_download(url)

    # Optional: compress after download
    compress = input("\n📦 Compress downloads into ZIP? (y/n): ").lower()
    if compress == "y":
        zip_folder("downloads")


if __name__ == "__main__":
    main()
```
