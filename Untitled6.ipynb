paused       = False
pause_lock   = threading.Lock()


# ══════════════════════════════════════════════════
#  PAUSE / RESUME  (background keyboard listener)
# ══════════════════════════════════════════════════
def _toggle_pause():
    global paused
    with pause_lock:
        paused = not paused
    state = "⏸  PAUSED   — press P to resume" if paused else "▶️  RESUMED  — downloading..."
    print(f"\n{state}\n")


def _keyboard_listener():
    """Daemon thread: press P to pause/resume, Q to quit."""
    global paused
    try:
        if os.name == "nt":                    # Windows
            import msvcrt
            while True:
                if msvcrt.kbhit():
                    ch = msvcrt.getwch().lower()
                    if ch == "p":
                        _toggle_pause()
                    elif ch == "q":
                        print("\n👋  Quit requested.\n")
                        os._exit(0)
                time.sleep(0.2)
        else:                                   # Linux / macOS
            import tty, termios, select
            fd  = sys.stdin.fileno()
            old = termios.tcgetattr(fd)
            try:
                tty.setraw(fd)
                while True:
                    r, _, _ = select.select([sys.stdin], [], [], 0.2)
                    if r:
                        ch = sys.stdin.read(1).lower()
                        if ch == "p":
                            _toggle_pause()
                        elif ch == "q":
                            print("\n👋  Quit requested.\n")
                            os._exit(0)
            finally:
                termios.tcsetattr(fd, termios.TCSADRAIN, old)
    except Exception:
        pass  # silently ignore if stdin unavailable


def start_keyboard_listener():
    t = threading.Thread(target=_keyboard_listener, daemon=True)
    t.start()


def wait_if_paused():
    """Blocks execution while paused. Call inside download loops."""
    while True:
        with pause_lock:
            if not paused:
                break
        time.sleep(0.5)


# ══════════════════════════════════════════════════
#  INTERNET CHECK & AUTO-RECONNECT
# ══════════════════════════════════════════════════
def is_online(host="8.8.8.8", port=53, timeout=3):
    """Ping Google DNS to check connectivity."""
    try:
        socket.setdefaulttimeout(timeout)
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.connect((host, port))
        s.close()
        return True
    except OSError:
        return False


def wait_for_internet(retry_secs=5, max_wait=600):
    """
    Block until internet comes back.
    Shows live countdown. Returns True when online, False on timeout.
    """
    if is_online():
        return True
    print("\n🔌  Internet disconnected — waiting to reconnect...")
    elapsed = 0
    while elapsed < max_wait:
        for remaining in range(retry_secs, 0, -1):
            print(f"\r   Retrying in {remaining}s  (total waited: {elapsed}s)   ", end="", flush=True)
            time.sleep(1)
        elapsed += retry_secs
        if is_online():
            print("\n✅  Internet restored — resuming!\n")
            return True
    print(f"\n❌  No internet after {max_wait}s — giving up.")
    return False


# ══════════════════════════════════════════════════
#  DOWNLOAD HISTORY
# ══════════════════════════════════════════════════
def load_history():
    if os.path.exists(HISTORY_FILE):
        try:
            with open(HISTORY_FILE, "r", encoding="utf-8") as f:
                return json.load(f)
        except Exception:
            pass
    return []


def save_history(entry: dict):
    history = load_history()
    history.append(entry)
    with open(HISTORY_FILE, "w", encoding="utf-8") as f:
        json.dump(history, f, indent=2, ensure_ascii=False)


def show_history():
    history = load_history()
    if not history:
        print("\n📭  No download history yet.")
        return
    print(f"\n📜  Download History  ({len(history)} total — showing last 30)")
    print("─" * 70)
    for i, e in enumerate(reversed(history[-30:]), 1):
        icon  = "✅" if e.get("status") == "ok" else "❌"
        ts    = e.get("time", "")[:16]
        title = e.get("title", e.get("url", "?"))[:50]
        print(f"  {i:>3}. {icon}  [{ts}]  {title}")
    print("─" * 70)

def safe_filename(name):
    for ch in '\\/*?:"<>|':
        name = name.replace(ch, "_")
    return name.strip()


def make_folder(base, sub=None):
    now  = datetime.now()
    path = Path(base) / str(now.year) / f"{now.month:02d}"
    if sub:
        path /= sub
    path.mkdir(parents=True, exist_ok=True)
    return str(path)


def zip_folder(folder):
    zip_name = f"{folder.rstrip(os.sep)}.zip"
    with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zf:
        for root, _, files in os.walk(folder):
            for file in files:
                full = os.path.join(root, file)
                zf.write(full, os.path.relpath(full, folder))
    print(f"📦  Compressed → {zip_name}")


def detect_content_type(url):
    u = url.lower()
    if any(x in u for x in ["youtube", "youtu.be", "tiktok", "vimeo",
                              "facebook", "x.com", "twitter", "instagram", "reddit"]):
        return "video"
    if any(x in u for x in ["soundcloud", "bandcamp"]):
        return "audio"
    if u.endswith((".jpg", ".jpeg", ".png", ".gif", ".webp")):
        return "image"
    if u.endswith(".pdf"):
        return "pdf"
    if u.endswith((".zip", ".rar", ".7z")):
        return "archive"
    if u.endswith((".doc", ".docx", ".ppt", ".pptx", ".xls", ".xlsx")):
        return "document"
    return "unknown"


def smart_folder(content_type):
    if content_type in ("video", "playlist"):
        return make_folder("downloads", "videos")
    if content_type == "audio":
        return make_folder("downloads", "audio")
    return make_folder("downloads", "files")

# ══════════════════════════════════════════════════
#  BUILD yt-dlp OPTIONS
# ══════════════════════════════════════════════════
def build_opts(folder, resolution="1080", audio_only=False,
               audio_quality="192", proxy=None, limit_rate=None,
               subtitles=False, playlist=True):

    if audio_only:
        fmt = "bestaudio/best"
    elif resolution == "best":
        fmt = "bestvideo+bestaudio/best"
    else:
        fmt = (
            f"bestvideo[height<={resolution}][ext=mp4]+bestaudio[ext=m4a]"
            f"/bestvideo[height<={resolution}]+bestaudio"
            f"/best[height<={resolution}]"
            f"/bestvideo+bestaudio/best"
        )

    opts = {
        "outtmpl":             os.path.join(folder, "%(title)s [%(height)sp].%(ext)s"),
        "merge_output_format": "mp4",
        "format":              fmt,

        # ── 403 / SABR fix ─────────────────────────────────────────
        "extractor_args":      {"youtube": {"player_client": ["android", "web"]}},

        # ── Resilience ─────────────────────────────────────────────
        "retries":             15,
        "fragment_retries":    15,
        "continuedl":          True,   # resume partial file
        "ignoreerrors":        True,
        "noplaylist":          not playlist,

        # ── Metadata ───────────────────────────────────────────────
        "writethumbnail":      True,
        "writeinfojson":       True,
        "write_description":   True,
        "writesubtitles":      subtitles,
        "writeautomaticsub":   subtitles,
        "subtitleslangs":      ["en", "hi"] if subtitles else [],

        # ── Output ─────────────────────────────────────────────────
        "quiet":               False,
        "progress":            True,

        # ── Post-processing ────────────────────────────────────────
        "postprocessors": [
            {"key": "FFmpegMetadata"},
            {"key": "EmbedThumbnail"},
        ],
    }

    if proxy:
        opts["proxy"] = proxy
    if limit_rate:
        opts["ratelimit"] = limit_rate

    if audio_only:
        opts["postprocessors"].append({
            "key":              "FFmpegExtractAudio",
            "preferredcodec":   "mp3",
            "preferredquality": audio_quality,
        })

    return opts

# Keywords that mean "network is gone" — not a format/auth error
_NET_ERRORS = [
    "network is unreachable", "name or service not known",
    "failed to resolve", "errno -2", "errno 101", "errno 110",
    "timed out", "read timed out", "connection reset",
    "connection refused", "broken pipe", "remote end closed",
    "10054", "10060", "10061", "nodename nor servname",
    "temporary failure in name resolution",
]

def _is_network_error(text: str) -> bool:
    t = text.lower()
    return any(kw in t for kw in _NET_ERRORS)


def _make_progress_hook():
    """
    Returns a yt-dlp progress hook that:
      1. Honours the global pause flag (blocks inside the hook)
      2. Detects a stalled/errored fragment and waits for internet
         BEFORE yt-dlp exhausts its own retry count.
    """
    last_bytes  = [0]
    last_time   = [time.time()]
    stall_limit = 30   # seconds with 0 new bytes → treat as disconnect

    def hook(d):
        wait_if_paused()

        if d["status"] == "downloading":
            downloaded = d.get("downloaded_bytes", 0)
            now        = time.time()

            if downloaded != last_bytes[0]:
                last_bytes[0] = downloaded
                last_time[0]  = now
            else:
                # No progress — check if stalled
                stalled_for = now - last_time[0]
                if stalled_for >= stall_limit and not is_online():
                    print(f"\n🔌  Stall detected ({stalled_for:.0f}s no data) — internet disconnected")
                    wait_for_internet(retry_secs=5)
                    last_time[0] = time.time()   # reset stall clock after reconnect

        elif d["status"] == "error":
            err = str(d.get("error", ""))
            if _is_network_error(err):
                print(f"\n🔌  Fragment error — waiting for internet...")
                wait_for_internet(retry_secs=5)

    return hook


def _do_download(url, opts, max_reconnect=999):
    """
    Wraps yt_dlp download with:
      • progress hook → pauses yt-dlp mid-download when internet drops
      • outer retry loop → restarts the entire download if yt-dlp gives up
      • continuedl=True  → always resumes the partial .mp4/.part file
      • unlimited reconnect attempts (max_reconnect=999 by default)

    This means:
      - Internet drops mid-download  → hook detects it, waits, yt-dlp resumes
      - yt-dlp exhausts its retries  → outer loop waits for internet, restarts
      - Partial file is NEVER deleted → always continues from last byte

    Returns ('ok', title) or ('fail', error_msg)
    """
    title = url

    # Inject our progress hook (keep any existing hooks the caller added)
    existing_hooks = opts.get("progress_hooks", [])
    opts["progress_hooks"] = existing_hooks + [_make_progress_hook()]

    # Force yt-dlp retries high so the hook has time to wait for reconnect
    opts["retries"]          = 99
    opts["fragment_retries"] = 99
    opts["continuedl"]       = True   # NEVER restart from zero

    attempt = 0
    while attempt <= max_reconnect:
        wait_if_paused()

        # Hard-check before starting/restarting
        if not is_online():
            print(f"\n🔌  No internet — waiting before attempt {attempt + 1}...")
            if not wait_for_internet():
                return ("fail", "No internet after maximum wait time")

        try:
            with yt_dlp.YoutubeDL(opts) as ydl:
                info = ydl.extract_info(url, download=True)
                if info:
                    title = info.get("title", url)
            return ("ok", title)   # ← success

        except yt_dlp.utils.DownloadError as e:
            err = str(e)
            if _is_network_error(err):
                attempt += 1
                print(f"\n🔌  yt-dlp gave up after retries (disconnect #{attempt})")
                print(   "    Waiting for internet then resuming partial file...\n")
                wait_for_internet(retry_secs=8)
                # Loop back — continuedl=True means it picks up the .part file
                continue
            # Non-network error (e.g. geo-block, bad format) — don't retry
            return ("fail", err)

        except Exception as e:
            err = str(e)
            if _is_network_error(err):
                attempt += 1
                wait_for_internet(retry_secs=8)
                continue
            return ("fail", err)

    return ("fail", f"Gave up after {max_reconnect} disconnect recoveries")

def download_yt_content(url, content_type, proxy=None, limit_rate=None):
    folder    = smart_folder(content_type)
    subtitles = ask_subtitles()

    if content_type == "audio":
        aq   = ask_audio_quality()
        opts = build_opts(folder, audio_only=True, audio_quality=aq,
                          proxy=proxy, limit_rate=limit_rate, subtitles=subtitles)
        res  = "audio"
    else:
        if input("\n👀  Show available formats first? (y/n): ").strip().lower() == "y":
            list_formats(url)
        res  = ask_quality()
        aq   = ask_audio_quality()
        opts = build_opts(folder, resolution=res, audio_quality=aq,
                          proxy=proxy, limit_rate=limit_rate, subtitles=subtitles)

    print(f"\n⏬  {content_type.upper()}  |  quality: {res}  |  → {folder}")
    print(   "   Press  P  to pause/resume   |   Q  to quit\n")

    status, result = _do_download(url, opts)

    save_history({
        "time":       datetime.now().isoformat(timespec="seconds"),
        "url":        url,
        "title":      result if status == "ok" else url,
        "type":       content_type,
        "resolution": res,
        "status":     status,
        "folder":     folder,
    })

    if status == "ok":
        print(f"\n✅  Done!  Saved → {folder}")
    else:
        print(f"\n❌  Failed: {result}")
        print("💡  pip install -U yt-dlp  then try again")


# ══════════════════════════════════════════════════
#  GENERIC FILE  (PDF / image / doc)
# ══════════════════════════════════════════════════
def generic_file_download(url, max_retries=5):
    folder  = smart_folder("files")
    headers = {"User-Agent": "Mozilla/5.0"}
    print(f"\n📥  {url}")

    for attempt in range(1, max_retries + 1):
        wait_if_paused()
        if not is_online():
            wait_for_internet()
        try:
            r     = requests.get(url, stream=True, timeout=30, headers=headers)
            if r.status_code != 200:
                print(f"❌  HTTP {r.status_code}")
                return
            ctype    = r.headers.get("content-type", "")
            ext      = ctype.split("/")[-1].split(";")[0] if "/" in ctype else "bin"
            fname    = safe_filename(url.split("/")[-1].split("?")[0]) or f"file_{int(time.time())}.{ext}"
            filepath = os.path.join(folder, fname)
            total    = int(r.headers.get("content-length", 0))
            done     = 0
            start    = time.time()
            with open(filepath, "ab") as f:
                for chunk in r.iter_content(65536):
                    wait_if_paused()
                    f.write(chunk)
                    done += len(chunk)
                    if total:
                        pct     = done / total * 100
                        elapsed = max(time.time() - start, 0.001)
                        speed   = done / elapsed / 1024
                        print(f"\r  {pct:5.1f}%  {done//1024}KB/{total//1024}KB  {speed:.0f}KB/s  ", end="", flush=True)
            print(f"\n✅  Saved → {filepath}")
            save_history({"time": datetime.now().isoformat(timespec="seconds"),
                          "url": url, "status": "ok", "folder": folder})
            return
        except requests.exceptions.ConnectionError:
            print(f"\n🔌  Connection lost (attempt {attempt}/{max_retries})")
            wait_for_internet()
        except Exception as e:
            print(f"\n⚠️  {e}")
            return
    print("❌  Max retries reached.")


# ══════════════════════════════════════════════════
#  BATCH MODE  (parallel)
# ══════════════════════════════════════════════════
def _batch_worker(args):
    url, opts, ctype = args
    if ctype in ("video", "audio"):
        return (url, *_do_download(url, opts))
    else:
        generic_file_download(url)
        return (url, "ok", url)


def batch_mode(file_path="urls.txt"):
    if not os.path.exists(file_path):
        print(f"⚠️  File not found: {file_path}")
        print("    Create it with one URL per line. Lines starting with # are skipped.")
        return

    with open(file_path, "r", encoding="utf-8") as f:
        urls = [l.strip() for l in f if l.strip() and not l.startswith("#")]

    if not urls:
        print("⚠️  No URLs found.")
        return

    print(f"\n📦  Batch mode — {len(urls)} URLs")
    res        = ask_quality()
    aq         = ask_audio_quality()
    subtitles  = ask_subtitles()
    proxy      = ask_proxy()
    limit_rate = ask_speed_limit()
    workers    = ask_workers()

    tasks = []
    for url in urls:
        ctype  = detect_content_type(url)
        folder = smart_folder(ctype)
        opts   = build_opts(folder, resolution=res, audio_only=(ctype == "audio"),
                            audio_quality=aq, proxy=proxy, limit_rate=limit_rate,
                            subtitles=subtitles)
        tasks.append((url, opts, ctype))

    print(f"\n⚡  Starting {workers} parallel workers...")
    print("   Press  P  to pause/resume  |  Q  to quit\n")

    ok_count = fail_count = 0
    with ThreadPoolExecutor(max_workers=workers) as pool:
        futures = {pool.submit(_batch_worker, t): t[0] for t in tasks}
        for future in as_completed(futures):
            url = futures[future]
            try:
                _, status, info = future.result()
                if status == "ok":
                    ok_count += 1
                    print(f"  ✅  {str(info)[:60]}")
                else:
                    fail_count += 1
                    print(f"  ❌  {url[:50]}  →  {str(info)[:40]}")
                save_history({"time": datetime.now().isoformat(timespec="seconds"),
                              "url": url, "title": str(info), "status": status})
            except Exception as e:
                fail_count += 1
                print(f"  ⚠️  {url[:50]}  →  {e}")

    print(f"\n═══ Batch done ═══  ✅ {ok_count} succeeded   ❌ {fail_count} failed")
    if input("\n📦  Compress all downloads to ZIP? (y/n): ").strip().lower() == "y":
        zip_folder("downloads")


# ══════════════════════════════════════════════════
#  AUTO-DETECT METHOD 1 — CLIPBOARD MONITOR
#  Keeps watching clipboard. When you copy a video
#  URL, it asks "Download this?" — press Enter = yes
# ══════════════════════════════════════════════════
def _get_clipboard() -> str:
    """Read clipboard text cross-platform (no extra lib needed)."""
    try:
        if os.name == "nt":                        # Windows
            import subprocess
            r = subprocess.run(
                ["powershell", "-command", "Get-Clipboard"],
                capture_output=True, text=True, timeout=3
            )
            return r.stdout.strip()
        else:                                       # Linux / macOS
            import subprocess
            # Try xclip, xsel, wl-paste, pbpaste in order
            for cmd in [["xclip", "-selection", "clipboard", "-o"],
                        ["xsel", "--clipboard", "--output"],
                        ["wl-paste", "--no-newline"],
                        ["pbpaste"]]:
                try:
                    r = subprocess.run(cmd, capture_output=True, text=True, timeout=3)
                    if r.returncode == 0:
                        return r.stdout.strip()
                except FileNotFoundError:
                    continue
    except Exception:
        pass
    return ""


VIDEO_DOMAINS = [
    "youtube.com", "youtu.be", "tiktok.com", "vimeo.com",
    "facebook.com", "instagram.com", "twitter.com", "x.com",
    "reddit.com", "soundcloud.com", "bandcamp.com",
]

def _looks_like_media_url(text: str) -> bool:
    if not text.startswith(("http://", "https://")):
        return False
    return any(d in text.lower() for d in VIDEO_DOMAINS)


def clipboard_monitor_mode():
    """
    Watches clipboard every second.
    When a new video URL is copied → asks user to confirm download.
    Press Ctrl+C to stop watching.
    """
    print("\n📋  CLIPBOARD MONITOR  — running")
    print("    Copy any video/audio URL and it will be detected automatically.")
    print("    Press  Ctrl+C  to stop.\n")

    seen      = set()
    last_text = _get_clipboard()   # seed with current clipboard so we don't re-trigger

    try:
        while True:
            time.sleep(1)
            text = _get_clipboard()
            if text and text != last_text and text not in seen:
                last_text = text
                if _looks_like_media_url(text):
                    seen.add(text)
                    print(f"\n🔗  Detected URL: {text[:80]}")
                    ctype = detect_content_type(text)
                    print(f"🧠  Type: {ctype.upper()}")
                    confirm = input("⬇️   Download this? (Enter=yes / n=skip): ").strip().lower()
                    if confirm != "n":
                        download_yt_content(text, ctype)
                    print("\n📋  Watching clipboard...")
    except KeyboardInterrupt:
        print("\n⏹  Clipboard monitor stopped.")


# ══════════════════════════════════════════════════
#  AUTO-DETECT METHOD 2 — WATCH FOLDER
#  Drop any .txt file with URLs into the watch
#  folder → script auto-downloads all links in it
# ══════════════════════════════════════════════════
WATCH_FOLDER = "watch_queue"

def watch_folder_mode():
    """
    Monitors the  watch_queue/  folder.
    Drop a .txt file with one URL per line → auto-downloads everything,
    then moves the file to  watch_queue/done/  so it won't re-trigger.
    Press Ctrl+C to stop.
    """
    queue_dir = Path(WATCH_FOLDER)
    done_dir  = queue_dir / "done"
    queue_dir.mkdir(exist_ok=True)
    done_dir.mkdir(exist_ok=True)

    print(f"\n📂  WATCH FOLDER MODE  — monitoring:  {queue_dir.resolve()}")
    print(    "    Drop a .txt file with one URL per line to auto-download.")
    print(    "    Processed files move to  watch_queue/done/")
    print(    "    Press  Ctrl+C  to stop.\n")

    processed = set()

    try:
        while True:
            time.sleep(2)
            for txt_file in queue_dir.glob("*.txt"):
                if txt_file in processed:
                    continue
                processed.add(txt_file)
                print(f"\n📄  New file detected: {txt_file.name}")
                try:
                    urls = [l.strip() for l in txt_file.read_text(encoding="utf-8").splitlines()
                            if l.strip() and not l.startswith("#")]
                except Exception as e:
                    print(f"⚠️  Could not read file: {e}")
                    continue

                if not urls:
                    print("   (empty file — skipping)")
                else:
                    print(f"   Found {len(urls)} URL(s) — starting download...\n")
                    res = ask_quality()
                    aq  = ask_audio_quality()
                    for url in urls:
                        ctype  = detect_content_type(url)
                        folder = smart_folder(ctype)
                        opts   = build_opts(folder, resolution=res,
                                            audio_only=(ctype == "audio"),
                                            audio_quality=aq)
                        print(f"  ⏬  {url[:70]}")
                        status, result = _do_download(url, opts)
                        icon = "✅" if status == "ok" else "❌"
                        print(f"  {icon}  {str(result)[:60]}")
                        save_history({"time": datetime.now().isoformat(timespec="seconds"),
                                      "url": url, "title": str(result), "status": status})

                # Move processed file to done/
                dest = done_dir / txt_file.name
                txt_file.rename(dest)
                print(f"\n📁  Moved to done: {dest}")
                print("📂  Watching for new files...")

    except KeyboardInterrupt:
        print("\n⏹  Watch folder stopped.")


# ══════════════════════════════════════════════════
#  AUTO-DETECT METHOD 3 — IMPROVED BATCH (urls.txt)
#  Now supports:
#  • inline comments after URL  (url  # my note)
#  • quality override per line  (url  !!720)
#  • skip already-downloaded    (checks history)
# ══════════════════════════════════════════════════
def _parse_urls_file(file_path: str):
    """
    Parse urls.txt with extended syntax:
      https://youtube.com/...             ← plain URL
      https://youtube.com/... # my note   ← URL + comment (ignored)
      https://youtube.com/... !!720       ← URL + quality override
      # this whole line is a comment
    Returns list of (url, quality_override_or_None)
    """
    entries = []
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"):
                continue
            quality = None
            # Strip inline comment
            if " #" in line:
                line = line.split(" #")[0].strip()
            # Check for quality override  !!720
            if " !!" in line:
                parts   = line.split(" !!")
                line    = parts[0].strip()
                quality = parts[1].strip()
            entries.append((line, quality))
    return entries


def _already_downloaded(url: str) -> bool:
    """Check download_history.json — skip if already succeeded."""
    for entry in load_history():
        if entry.get("url") == url and entry.get("status") == "ok":
            return True
    return False


def batch_mode(file_path="urls.txt"):
    if not os.path.exists(file_path):
        print(f"⚠️  File not found: {file_path}")
        print("    Create it with one URL per line.")
        print("    Tip: add  !!720  after a URL to override quality for that video.")
        print("    Example:  https://youtube.com/watch?v=abc  !!720")
        return

    entries = _parse_urls_file(file_path)
    if not entries:
        print("⚠️  No URLs found.")
        return

    # Skip already downloaded
    fresh = [(u, q) for u, q in entries if not _already_downloaded(u)]
    skipped = len(entries) - len(fresh)
    if skipped:
        print(f"⏭   Skipping {skipped} already-downloaded URL(s).")
    if not fresh:
        print("✅  All URLs already downloaded!")
        return

    print(f"\n📦  Batch mode — {len(fresh)} URL(s)  ({skipped} skipped)\n")

    default_res = ask_quality()
    aq          = ask_audio_quality()
    subtitles   = ask_subtitles()
    proxy       = ask_proxy()
    limit_rate  = ask_speed_limit()
    workers     = ask_workers()

    tasks = []
    for url, quality_override in fresh:
        ctype  = detect_content_type(url)
        folder = smart_folder(ctype)
        res    = quality_override or default_res
        opts   = build_opts(folder, resolution=res, audio_only=(ctype == "audio"),
                            audio_quality=aq, proxy=proxy, limit_rate=limit_rate,
                            subtitles=subtitles)
        tasks.append((url, opts, ctype))

    print(f"\n⚡  Starting {workers} parallel workers...")
    print("   Press  P  to pause/resume  |  Q  to quit\n")

    ok_count = fail_count = 0
    with ThreadPoolExecutor(max_workers=workers) as pool:
        futures = {pool.submit(_batch_worker, t): t[0] for t in tasks}
        for future in as_completed(futures):
            url = futures[future]
            try:
                _, status, info = future.result()
                if status == "ok":
                    ok_count += 1
                    print(f"  ✅  {str(info)[:65]}")
                else:
                    fail_count += 1
                    print(f"  ❌  {url[:50]}  →  {str(info)[:40]}")
                save_history({"time": datetime.now().isoformat(timespec="seconds"),
                              "url": url, "title": str(info), "status": status})
            except Exception as e:
                fail_count += 1
                print(f"  ⚠️  {url[:50]}  →  {e}")

    print(f"\n═══ Batch done ═══  ✅ {ok_count} succeeded   ❌ {fail_count} failed")
    if input("\n📦  Compress all downloads to ZIP? (y/n): ").strip().lower() == "y":
        zip_folder("downloads")

def _is_jupyter() -> bool:
    """
    Returns True if running inside Jupyter notebook / JupyterLab.
    In Jupyter, sys.argv contains kernel flags like -f, --ip etc.
    — we must NOT use sys.argv for our own argument parsing.
    """
    try:
        shell = get_ipython().__class__.__name__   # type: ignore[name-defined]
        return shell in ("ZMQInteractiveShell", "TerminalInteractiveShell")
    except NameError:
        return False


def main():
    start_keyboard_listener()   # P = pause/resume, Q = quit

    # ── Jupyter guard ─────────────────────────────────────────────
    # sys.argv inside Jupyter = Jupyter's own kernel flags, not ours.
    # Skip all CLI parsing and go straight to interactive menu.
    if _is_jupyter():
        print("📓  Jupyter detected — opening interactive menu.\n")
        interactive_menu()
        return
    # ──────────────────────────────────────────────────────────────

    if len(sys.argv) > 1:
        cmd = sys.argv[1]
        if cmd == "--batch":
            batch_mode(sys.argv[2] if len(sys.argv) > 2 else "urls.txt")
        elif cmd == "--clipboard":
            clipboard_monitor_mode()
        elif cmd == "--watch":
            watch_folder_mode()
        elif cmd == "--history":
            show_history()
        elif cmd == "--formats":
            url = sys.argv[2] if len(sys.argv) > 2 else input("🔗  URL: ").strip()
            list_formats(url)
        elif cmd == "--help":
            print(__doc__)
        elif cmd.startswith("http"):
            # Direct URL as terminal argument
            url   = cmd
            ctype = detect_content_type(url)
            print(f"🧠  Detected: {ctype.upper()}")
            if ctype in ("video", "audio"):
                download_yt_content(url, ctype)
            else:
                generic_file_download(url)
        else:
            # Unknown flag (e.g. Jupyter's -f leaking through) → open menu
            print(f"⚠️  Unknown argument '{cmd}' — opening menu instead.")
            interactive_menu()
    else:
        interactive_menu()


if __name__ == "__main__":
    main()
