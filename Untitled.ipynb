```python
import yt_dlp
import requests
import aiohttp
import asyncio
import os
import sys
import json
import re
import time
import hashlib
import zipfile
import schedule
import threading
from pathlib import Path
from datetime import datetime
from tqdm import tqdm
from rich.console import Console

console = Console()

# ─────────────────────────────────────────────
#  Utility & System Functions
# ─────────────────────────────────────────────
def safe_filename(name):
    """Clean up filenames intelligently using AI-like cleanup."""
    name = re.sub(r'[_\-]+', ' ', name)
    name = re.sub(r'\s+', ' ', name).strip()
    name = re.sub(r'[^A-Za-z0-9 .()]+', '', name)
    if len(name) > 70:
        name = name[:70] + "..."
    return name.title()

def make_folder(base, sub=None):
    now = datetime.now()
    folder = Path(base) / str(now.year) / f"{now.month:02d}"
    if sub:
        folder /= sub
    folder.mkdir(parents=True, exist_ok=True)
    return folder

def sha256sum(filename):
    h = hashlib.sha256()
    with open(filename, "rb") as f:
        for chunk in iter(lambda: f.read(8192), b""):
            h.update(chunk)
    return h.hexdigest()

# ─────────────────────────────────────────────
#  Detect File Type
# ─────────────────────────────────────────────
def detect_content_type(url):
    u = url.lower()
    if any(x in u for x in ["youtube", "youtu.be", "tiktok", "vimeo", "facebook", "instagram", "x.com", "reddit"]):
        return "video"
    if any(x in u for x in ["soundcloud", "bandcamp", "spotify"]):
        return "audio"
    if u.endswith((".jpg", ".jpeg", ".png", ".gif", ".webp")):
        return "image"
    if u.endswith(".pdf"):
        return "pdf"
    if u.endswith((".zip", ".rar", ".7z")):
        return "archive"
    if u.endswith((".doc", ".docx", ".ppt", ".pptx", ".xls", ".xlsx")):
        return "document"
    return "unknown"

# ─────────────────────────────────────────────
#  Generic File Downloader
# ─────────────────────────────────────────────
async def download_generic_file(url, folder):
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as resp:
            if resp.status != 200:
                console.print(f"[red]Failed to download: {url}[/red]")
                return
            content_type = resp.headers.get("content-type", "")
            ext = content_type.split('/')[-1] if '/' in content_type else "bin"
            filename = safe_filename(url.split("/")[-1].split("?")[0]) or f"file_{int(time.time())}.{ext}"
            filepath = folder / filename
            with open(filepath, "wb") as f:
                async for chunk in resp.content.iter_chunked(8192):
                    f.write(chunk)
            console.print(f"[green]✅ Saved → {filepath}[/green]")
            log_integrity(filepath)

# ─────────────────────────────────────────────
#  yt-dlp Video/Audio Downloader
# ─────────────────────────────────────────────
def yt_download(url, folder, content_type):
    opts = {
        "outtmpl": str(folder / "%(title)s.%(ext)s"),
        "writethumbnail": True,
        "writeinfojson": True,
        "writesubtitles": True,
        "write_description": True,
        "continuedl": True,
        "merge_output_format": "mp4",
        "retries": 5,
        "quiet": False,
        "ignoreerrors": True,
        "progress": True,
    }

    if content_type == "audio":
        opts["format"] = "bestaudio/best"
        opts["postprocessors"] = [
            {"key": "FFmpegExtractAudio", "preferredcodec": "mp3", "preferredquality": "192"},
            {"key": "FFmpegMetadata"},
        ]
    else:
        opts["format"] = "bestvideo+bestaudio/best"

    console.print(f"[cyan]⏬ Downloading {content_type} → {url}[/cyan]")
    with yt_dlp.YoutubeDL(opts) as ydl:
        ydl.download([url])
    console.print(f"[green]✅ {content_type.capitalize()} saved in {folder}[/green]")

# ─────────────────────────────────────────────
#  Integrity & ZIP
# ─────────────────────────────────────────────
def log_integrity(filepath):
    h = sha256sum(filepath)
    log_entry = {"file": str(filepath), "sha256": h, "time": time.ctime()}
    with open("integrity.json", "a") as f:
        f.write(json.dumps(log_entry) + "\n")

def zip_day_folder():
    now = datetime.now()
    base = Path("downloads") / str(now.year) / f"{now.month:02d}"
    zip_name = f"downloads_{now.strftime('%Y%m%d')}.zip"
    with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as z:
        for root, _, files in os.walk(base):
            for file in files:
                full = os.path.join(root, file)
                z.write(full, os.path.relpath(full, base))
    console.print(f"[blue]📦 Auto-Zipped → {zip_name}[/blue]")

# ─────────────────────────────────────────────
#  Batch Download (Multi-threaded + Async)
# ─────────────────────────────────────────────
async def batch_download(urls):
    tasks = []
    for url in urls:
        ctype = detect_content_type(url)
        folder = make_folder("downloads", ctype)
        if ctype in ["video", "audio"]:
            loop = asyncio.get_event_loop()
            await loop.run_in_executor(None, yt_download, url, folder, ctype)
        else:
            tasks.append(download_generic_file(url, folder))
    if tasks:
        await asyncio.gather(*tasks)
    zip_day_folder()

# ─────────────────────────────────────────────
#  AI Filename Cleanup on Old Files
# ─────────────────────────────────────────────
def ai_cleanup_filenames():
    base = Path("downloads")
    for path in base.rglob("*.*"):
        clean_name = safe_filename(path.stem) + path.suffix
        new_path = path.parent / clean_name
        if new_path != path:
            path.rename(new_path)
    console.print("[magenta]✨ AI filename cleanup completed.[/magenta]")

# ─────────────────────────────────────────────
#  Main CLI Logic
# ─────────────────────────────────────────────
def main():
    console.print("[bold cyan]⚡ OMNI DOWNLOADER Ω — AI Universal Engine[/bold cyan]\n")

    if len(sys.argv) > 1:
        arg = sys.argv[1].lower()
        if arg == "--batch":
            if not os.path.exists("urls.txt"):
                console.print("[red]No urls.txt found[/red]")
                return
            with open("urls.txt") as f:
                urls = [line.strip() for line in f if line.strip()]
            asyncio.run(batch_download(urls))
            return
        elif arg == "--ai-clean":
            ai_cleanup_filenames()
            return
        elif arg == "--zip":
            zip_day_folder()
            return

    url = input("🔗 Enter any URL: ").strip()
    if not url:
        console.print("[red]No URL entered[/red]")
        return

    ctype = detect_content_type(url)
    folder = make_folder("downloads", ctype)
    console.print(f"🧠 Detected: {ctype.upper()}")

    if ctype in ["video", "audio"]:
        yt_download(url, folder, ctype)
    else:
        asyncio.run(download_generic_file(url, folder))

    ans = input("📦 Zip today's downloads? (y/n): ").lower()
    if ans == "y":
        zip_day_folder()


if __name__ == "__main__":
    main()
```
