```python
# ══════════════════════════════════════════
# STEP 1 — Fix pip & Install (Run ONCE only)
# ══════════════════════════════════════════
import subprocess, sys

print("🔧 Fixing conda + installing yt-dlp...")
subprocess.run(["conda", "install", "-c", "conda-forge", "yt-dlp", "ffmpeg",
                "-y", "--no-update-deps"], check=False)
print("✅ Done! Now run STEP 2 below.")
```

```python
# ══════════════════════════════════════════════════════════════
#  ⚡ SMART DOWNLOADER — PASTE URL & RUN
#  Supports: Full Season, Playlist, Single Video, Movie
#  Auto-resume if internet cuts — never re-downloads same file
# ══════════════════════════════════════════════════════════════

import os, yt_dlp
from datetime import datetime

# ┌─────────────────────────────────────────┐
# │  ✏️  CHANGE THESE 3 LINES ONLY          │
URL        = "https://www.youtube.com/playlist?list=PLxxxxxx"  # ← PASTE YOUR REAL URL HERE
SHOW_NAME  = "My Show Season 1"   # ← folder name (anything you want)
QUALITY    = "3"                  # ← 1=4K  2=1440p  3=1080p  4=720p  5=480p  6=360p  7=best
# └─────────────────────────────────────────┘

# ── Validation ──
if "EXAMPLE" in URL or "PLxxxxxx" in URL or not URL.startswith("http"):
    print("❌ ERROR: You forgot to change the URL!")
    print("   Replace the URL above with your real video/playlist link.")
    raise SystemExit

# ── Setup folders ──
base      = os.path.join(os.path.expanduser("~"), "downloads")
date_dir  = datetime.now().strftime("%Y-%m")
safe_name = "".join(c if c.isalnum() or c in " _-" else "_" for c in SHOW_NAME).strip()
folder    = os.path.join(base, date_dir, safe_name)
os.makedirs(folder, exist_ok=True)
archive   = os.path.join(folder, "downloaded.txt")

print(f"📁 Save folder : {folder}")
print(f"📦 Archive     : {archive}  ← tracks what's already downloaded")
print(f"🎯 Quality     : {QUALITY}")
print(f"🔗 URL         : {URL}\n")

# ── Quality map ──
qmap = {
    "1": "bestvideo[height<=2160]+bestaudio/bestvideo+bestaudio/best",
    "2": "bestvideo[height<=1440]+bestaudio/bestvideo+bestaudio/best",
    "3": ("bestvideo[height<=1080]+bestaudio/"
          "bestvideo[height<=720]+bestaudio/"
          "bestvideo[height<=480]+bestaudio/best"),
    "4": "bestvideo[height<=720]+bestaudio/best",
    "5": "bestvideo[height<=480]+bestaudio/best",
    "6": "bestvideo[height<=360]+bestaudio/best",
    "7": "bestvideo+bestaudio/best",
}
fmt = qmap.get(QUALITY, QUALITY)

# ── Download options ──
opts = {
    "outtmpl"            : os.path.join(folder,
                               "%(playlist_index)s - %(title)s.%(ext)s"),
    "format"             : fmt,
    "merge_output_format": "mp4",

    # ✅ Auto-resume — never restart from zero
    "continuedl"         : True,
    "retries"            : 99,
    "fragment_retries"   : 99,
    "retry_sleep_functions": {"http": lambda n: 5},  # wait 5s between retries

    # ✅ Skip already downloaded
    "download_archive"   : archive,
    "nooverwrites"       : True,

    # ✅ Skip broken videos, keep downloading rest
    "ignoreerrors"       : True,

    "extractor_args"     : {"youtube": {"player_client": ["android", "web"]}},
    "http_headers"       : {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/120.0.0.0 Safari/537.36"
        )
    },
}

# ── Download with cookie fallback ──
def run(use_cookies):
    o = opts.copy()
    if use_cookies:
        o["cookiesfrombrowser"] = ("chrome",)
    with yt_dlp.YoutubeDL(o) as ydl:
        ydl.download([URL])

try:
    print("⏬ Starting download (with Chrome cookies)...\n")
    run(True)
except Exception as e:
    print(f"⚠️  Cookie mode failed: {e}")
    print("⏬ Retrying without cookies (public mode)...\n")
    try:
        run(False)
    except Exception as e2:
        print(f"❌ Failed: {e2}")
        raise SystemExit

print(f"\n✅ ALL DONE!")
print(f"📁 Files saved in: {folder}")
print(f"📋 Already-downloaded list: {archive}")
print(f"\n💡 If internet cut off — just run this cell AGAIN.")
print(f"   It will AUTO-RESUME and skip already finished files.")
```

```python
# ══════════════════════════════════════════════════════
#  🔍 OPTIONAL — See ALL available qualities & sizes
#  Run this BEFORE downloading to pick the right quality
# ══════════════════════════════════════════════════════

import yt_dlp

URL = "https://www.youtube.com/watch?v=PASTE_REAL_URL_HERE"  # ← paste URL here

if "PASTE_REAL" in URL or not URL.startswith("http"):
    print("❌ Change the URL above first!")
    raise SystemExit

print(f"🔍 Loading formats for:\n   {URL}\n")

with yt_dlp.YoutubeDL({"quiet": True, "skip_download": True}) as ydl:
    info = ydl.extract_info(URL, download=False)
    if "entries" in info:
        print(f"📋 Playlist: {info.get('title','')} — showing first video formats\n")
        info = next(iter(info["entries"]), None)
    if not info:
        print("❌ Could not get info.")
    else:
        print(f"🎬 {info.get('title','')}\n")
        print(f"{'ID':<12} {'EXT':<6} {'RESOLUTION':<12} {'FPS':<6} {'SIZE'}")
        print("─" * 55)
        seen = set()
        for f in info.get("formats", []):
            if f.get("vcodec","none") == "none": continue
            h   = f.get("height")
            res = f"{h}p" if h else "?"
            ext = f.get("ext","?")
            if (res,ext) in seen: continue
            seen.add((res,ext))
            fps  = f.get("fps") or "-"
            fs   = f.get("filesize") or f.get("filesize_approx")
            size = f"{fs/1024/1024:.1f} MB" if fs else "~unknown"
            fid  = f.get("format_id","?")
            print(f"{fid:<12} {ext:<6} {res:<12} {str(fps):<6} {size}")
        print("\n💡 Paste the ID into QUALITY= in the main cell above.")
```
