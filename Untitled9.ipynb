```python
### !/usr/bin/env python3
import os
import yt_dlp
from datetime import datetime

# ----------------------------
# BASIC SAFETY CHECKS
# ----------------------------

def is_valid_url(url: str) -> bool:
    return url.startswith("http") and ("watch?v=" in url or "youtu.be/" in url or "vimeo.com" in url)

def get_folder():
    path = f"downloads/{datetime.now().strftime('%Y-%m')}"
    os.makedirs(path, exist_ok=True)
    return path

# ----------------------------
# FORMAT BUILDER (SAFE)
# ----------------------------

def get_format(choice: str):
    formats = {
        "1": "bestvideo[height<=2160]+bestaudio/best",
        "2": "bestvideo[height<=1440]+bestaudio/best",
        "3": "bestvideo[height<=1080]+bestaudio/best",
        "4": "bestvideo[height<=720]+bestaudio/best",
        "5": "bestaudio/best"
    }
    return formats.get(choice, formats["3"])

# ----------------------------
# CORE DOWNLOAD ENGINE
# ----------------------------

def download(url: str, mode="video", quality="3"):
    if not is_valid_url(url):
        print("❌ Invalid URL. Please paste a real video link.")
        return

    folder = get_folder()

    if mode == "audio":
        fmt = "bestaudio/best"
    else:
        fmt = get_format(quality)

    ydl_opts = {
        "format": fmt,
        "outtmpl": f"{folder}/%(title)s.%(ext)s",
        "noplaylist": False,
        "retries": 5,
        "quiet": False,
        "merge_output_format": "mp4",
    }

    # audio conversion
    if mode == "audio":
        ydl_opts["postprocessors"] = [{
            "key": "FFmpegExtractAudio",
            "preferredcodec": "mp3",
            "preferredquality": "192",
        }]

    try:
        print("\n⏬ Starting download...\n")
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=True)

            # FIX: prevent fake success
            if not info:
                print("❌ No media found in URL.")
                return

            title = info.get("title", "Unknown")
            print(f"\n✅ Done: {title}")
            print(f"📁 Saved in: {folder}")

    except Exception as e:
        print(f"\n❌ Error: {str(e)}")

# ----------------------------
# FORMAT LIST
# ----------------------------

def list_formats(url):
    if not is_valid_url(url):
        print("❌ Invalid URL")
        return

    with yt_dlp.YoutubeDL({"listformats": True}) as ydl:
        ydl.download([url])

# ----------------------------
# SIMPLE MENU
# ----------------------------

def main():
    while True:
        print("""
==============================
⚡ Downloader v6 (Stable)
==============================
[1] Video Download
[2] Audio Download (MP3)
[3] List Formats
[0] Exit
""")

        choice = input("Select: ").strip()

        if choice == "0":
            break

        url = input("Paste URL: ").strip()

        if choice == "1":
            q = input("Quality (1=4K 2=1440 3=1080 4=720): ").strip()
            download(url, "video", q)

        elif choice == "2":
            download(url, "audio")

        elif choice == "3":
            list_formats(url)

        else:
            print("❌ Invalid option")

if __name__ == "__main__":
    main()
```
